## 두 수의 나눗셈

```python
# 정수 num1과 num2가 매개변수로 주어질 때, 
# num1을 num2로 나눈 값에 1,000을 곱한 후 정수 부분을 return 하도록
# solution 함수를 완성해주세요.
def solution(num1, num2):
    return int(num1 / num2 * 1000)

# 다른 사람 풀이
solution = lambda x, y : 1000 * x // y
```

```python
solution(3, 2)
```

## 숫자 비교하기

```python
# 정수 num1과 num2가 매개변수로 주어집니다. 두 수가 같으면 1 다르면 -1을 retrun하도록 solution 함수를 완성해주세요.
def solution(num1, num2):
    if num1 == num2:
        return 1
    return -1

# 다른 사람 풀이
def solution(num1, num2):
    return 1 if num1 == num2 else -1
```

```python
solution(2, 3)
```

## 분수의 덧셈

```python
# 첫 번째 분수의 분자와 분모를 뜻하는 numer1, denom1, 
# 두 번째 분수의 분자와 분모를 뜻하는 numer2, denom2가 매개변수로 주어집니다. 
# 두 분수를 더한 값을 기약 분수로 나타냈을 때 
# 분자와 분모를 순서대로 담은 배열을 return 하도록
# solution 함수를 완성해보세요.
def solution(numer1, denom1, numer2, denom2):  # 두 분수 numer1/denom1, numer2/denom2를 더하는 함수
    numer1 *= denom2                          # 첫 번째 분수의 분자에 두 번째 분모를 곱해 분모 통일
    numer2 *= denom1                          # 두 번째 분수의 분자에 첫 번째 분모를 곱해 분모 통일
    denominator = denom1 * denom2             # 두 분수의 공통 분모 생성

    numerator = numer1 + numer2               # 분모가 같아졌으므로 분자끼리 더함

    def gcd(a, b):                            # 두 수의 최대공약수를 구하는 함수 (유클리드 호제법)
        while b != 0:                         # 나머지가 0이 될 때까지 반복
            a, b = b, a % b                   # a를 b로, b를 a를 b로 나눈 나머지로 갱신
        return a                              # 반복이 끝나면 a가 최대공약수

    g = gcd(numerator, denominator)           # 분자와 분모의 최대공약수 계산
    numerator //= g                           # 분자를 최대공약수로 나누어 약분
    denominator //= g                         # 분모를 최대공약수로 나누어 약분

    return [numerator, denominator]           # 기약분수 형태로 결과 반환


# 다른 사람 풀이
import math

def solution(numer1, denom1, numer2, denom2):
    numerator = numer1 * denom2 + numer2 * denom1
    denominator = denom1 * denom2

    gcd = math.gcd(numerator, denominator)
    
    return [numerator // gcd, denominator // gcd]
```

```python
solution(9, 2, 1, 3)
```

## 배열 두 배 만들기

```python
# 정수 배열 numbers가 매개변수로 주어집니다.
# numbers의 각 원소에 두배한 원소를 가진 배열을 return하도록
# solution 함수를 완성해주세요.
def solution(numbers):
    lst = []
    for i in numbers:
        lst.append(i * 2)
    return lst

# 다른 사람 풀이
def solution(numbers):
    return [num*2 for num in numbers]
```

```python
solution([1,2,3,4,5])
```
