## 가까운 수

```python
# 정수 배열 array와 정수 n이 매개변수로 주어질 때, 
# array에 들어있는 정수 중 n과 가장 가까운 수를 
# return 하도록 solution 함수를 완성해주세요.
def solution(array, n):
    lst = []                 # n과 각 원소의 차이(절댓값)를 저장할 빈 리스트 생성
    
    array.sort()             # array를 오름차순으로 정렬
                              # → 거리가 같을 경우 더 작은 수가 먼저 나오게 하기 위함
    
    for i in array:          # array에 들어있는 각 숫자를 하나씩 순회
        lst.append(abs(n-i)) # n과 현재 숫자 i의 차이를 절댓값으로 구해서 lst에 저장 [17, 10, 8]
    
    return array[lst.index(min(lst))] # 8의 인덱스 구하고 -> array[2]니까 28
                              # lst에서 가장 작은 거리값을 찾고
                              # 그 값이 처음 등장하는 인덱스를 구한 뒤
                              # 같은 인덱스의 array 값을 반환

# GPT 풀이
def solution(array, n):
    return min(array, key=lambda x: (abs(x - n), x))
```

```python
solution([3, 10, 28], 20)
```

## 369 게임

```python
# 머쓱이는 친구들과 369게임을 하고 있습니다. 
# 369게임은 1부터 숫자를 하나씩 대며 3, 6, 9가 들어가는 숫자는 숫자 대신 3, 6, 9의 개수만큼 박수를 치는 게임입니다.
# 머쓱이가 말해야하는 숫자 order가 매개변수로 주어질 때, 
# 머쓱이가 쳐야할 박수 횟수를 return 하도록 solution 함수를 완성해보세요.
def solution(order):
    ans = 0
    for i in str(order):
        if i in ['3', '6', '9']:
            ans += 1
    return ans

# 다른 사람 풀이
def solution(order):
    return sum(map(lambda x: str(order).count(str(x)), [3, 6, 9]))

# 다른 사람 풀이
def solution(order):
    order = str(order)
    return order.count('3') + order.count('6') + order.count('9')
```

```python
solution(3)
```

## 암호해독

```python
# 군 전략가 머쓱이는 전쟁 중 적군이 다음과 같은 암호 체계를 사용한다는 것을 알아냈습니다.
# 암호화된 문자열 cipher를 주고받습니다.
# 그 문자열에서 code의 배수 번째 글자만 진짜 암호입니다.
# 문자열 cipher와 정수 code가 매개변수로 주어질 때
# 해독된 암호 문자열을 return하도록 solution 함수를 완성해주세요.
def solution(cipher, code):
    ans = ''
    for i in cipher[code-1::code]:
        ans += i
    return ans

# 다른 사람 풀이
def solution(cipher, code):
    answer = ''

    for i in range(code-1,len(cipher),code):
        answer += cipher[i]

    return answer
```

```python
solution("dfjardstddetckdaccccdegk", 4)
```

## 대문자와 소문자

```python
# 문자열 my_string이 매개변수로 주어질 때, 대문자는 소문자로 소문자는 대문자로 변환한 문자열을 return하도록 solution 함수를 완성해주세요.
def solution(my_string):
    return my_string.swapcase()

# 다른 사람 풀이
def solution(my_string):
    answer = ''

    for i in my_string:
        if i.isupper():
            answer+=i.lower()
        else:
            answer+=i.upper()
    return answer
```

```python
solution("cccCCC")
```
