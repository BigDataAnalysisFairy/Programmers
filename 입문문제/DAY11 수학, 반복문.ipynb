## 주사위의 개수

```python
# 머쓱이는 직육면체 모양의 상자를 하나 가지고 있는데
# 이 상자에 정육면체 모양의 주사위를 최대한 많이 채우고 싶습니다.
# 상자의 가로, 세로, 높이가 저장되어 있는 배열 box와
# 주사위 모서리의 길이 정수 n이 매개변수로 주어졌을 때,
# 상자에 들어갈 수 있는 주사위의 최대 개수를
# return 하도록 solution 함수를 완성해주세요.
def solution(box, n):
    ans = 1
    for i in box:
        ans *= (i // n) # 가로 세로 높이는 서로 독립이므로 곱해줌
    return ans

# 다른 사람 풀이
def solution(box, n):
    x, y, z = box
    return (x // n) * (y // n) * (z // n )
```

```python
solution([10,8,6], 3)
```

## 합성수 찾기

```python
# 약수의 개수가 세 개 이상인 수를 합성수라고 합니다.
# 자연수 n이 매개변수로 주어질 때 n이하의 합성수의 개수를 return하도록
# solution 함수를 완성해주세요.
def solution(n):
    ans = 0
    for i in range(1, n+1):
        total = 0
        for j in range(1, i+1):
            if i % j == 0:
                total += 1
        if total >= 3:
            ans += 1
    return ans

# 다른 사람 풀이
# Dong-Yeon Park님께 답변드리자면 i까지로 하면 
# 컴파일 시간이 너무 길어지기 때문입니다! 반복 범위를 줄이기 위해,
# 대부분 자연수의 약수는 대상의 제곱근 이하의 약수 하나와
# 이상의 약수 하나씩 한 쌍을 이룬다는 점을 응용해
# i의 제곱근까지 반복하게 한 다음, 
# 9등의 제곱수를 고려해 +1을 해서 예외를 해결한 경우입니다!
# 4부터 최초의 합성수. 1, 2, 3은 애초에 검사할 필요 없음
def solution(n):                              # 1부터 n까지의 수 중에서 합성수의 개수를 구하는 함수
    output = 0                               # 합성수의 개수를 저장할 변수

    for i in range(4, n + 1):                # 4부터 n까지 검사 (4가 가장 작은 합성수)
        for j in range(2, int(i ** 0.5) + 1):# 2부터 √i까지 약수가 있는지 확인
                                              # (1은 항상 나누어지므로 검사할 필요 없음)
            if i % j == 0:                   # i가 j로 나누어 떨어지면
                output += 1                  # i는 합성수이므로 개수 1 증가
                break                        # 약수 하나만 발견해도 합성수 확정 → 더 검사하지 않음

    return output                            # 합성수의 총 개수 반환
```

```python
solution(10)
```

## 최댓값 만들기 (1)

```python
# 정수 배열 numbers가 매개변수로 주어집니다. 
# numbers의 원소 중 두 개를 곱해 만들 수 있는 최댓값을 return하도록 
# solution 함수를 완성해주세요.
def solution(numbers):
    numbers.sort()
    return numbers[-2] * numbers[-1]

# 다른 사람 풀이
def solution(numbers):
    return sorted(numbers)[-1] * sorted(numbers)[-2]
```

```python
solution([1, 2, 4, 3, 5])
```

## 팩토리얼

```python
# i팩토리얼 (i!)은 1부터 i까지 정수의 곱을 의미합니다.
# 예를들어 5! = 5 * 4 * 3 * 2 * 1 = 120 입니다. 
# 정수 n이 주어질 때 다음 조건을 만족하는 가장 큰 정수 i를 return 하도록 
# solution 함수를 완성해주세요.
def solution(n):                         # 주어진 n 이하에서 만들 수 있는 최대 팩토리얼의 값의 "숫자"를 구하는 함수
    fact = 1                            # 팩토리얼 값을 저장 (1! = 1부터 시작)
    i = 1                               # 현재 곱할 숫자 (1, 2, 3, 4, ...)

    while fact * i <= n:                # 다음 팩토리얼 값이 n을 넘지 않는 동안 반복
        fact *= i                       # fact에 i를 곱해 i! 계산
        i += 1                          # 다음 숫자로 증가

    return i - 1                        # 조건을 넘기 직전의 i가 최대 정답이므로 i-1 반환


# 다른 사람 풀이
from math import factorial

def solution(n):
    k = 10
    while n < factorial(k):
        k -= 1
    return k
```
