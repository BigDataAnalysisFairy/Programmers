## 편지

```python
# 머쓱이는 할머니께 생신 축하 편지를 쓰려고 합니다.
# 할머니가 보시기 편하도록 글자 한 자 한 자를 가로 2cm 크기로 적으려고 하며, 편지를 가로로만 적을 때,
# 축하 문구 message를 적기 위해 필요한 편지지의 최소 가로길이를 return 하도록 
# solution 함수를 완성해주세요.
def solution(message):
    return len(message) * 2

# 다른 사람 풀이
def solution(message):
    return len(message)<<1
```

```python
solution("happy birthday!")
```

## 가장 큰 수 찾기

```python
# 정수 배열 array가 매개변수로 주어질 때, 
# 가장 큰 수와 그 수의 인덱스를 담은 배열을 return 하도록
#  solution 함수를 완성해보세요.
def solution(array):
    lst = []
    lst.append(max(array))
    lst.append(array.index(max(array)))
    return lst

# 다른 사람 풀이
def solution(array):
    return [max(array), array.index(max(array))]
```

```python
solution([1, 8, 3])
```

## 문자열 계산하기

```python
# my_string은 "3 + 5"처럼 문자열로 된 수식입니다. 문자열 my_string이 매개변수로 주어질 때, 수식을 계산한 값을 return 하는 solution 함수를 완성해주세요.

def solution(my_string):
    return eval(my_string)

# 재풀이
def solution(my_string):
    ans = 0
    for i in my_string.replace(' - ', ' + -').split(' + '):
        ans += int(i)
    return ans

# 다른 사람 풀이
def solution(my_string):
    return sum(int(i) for i in my_string.replace(' - ', ' + -').split(' + '))
```

```python
solution("3 + 4")
```

## 배열의 유사도

```python
# 두 배열이 얼마나 유사한지 확인해보려고 합니다. 
# 문자열 배열 s1과 s2가 주어질 때 같은 원소의 개수를
# return하도록 solution 함수를 완성해주세요.
def solution(s1, s2):
    return len(set(s1) & set(s2))

# 다른 사람 풀이
def solution(s1, s2):
    answer = 0

    for word in s1:
        if word in s2:
            answer += 1
        else:
            continue

    return answer
```

```python
solution(["n", "omg"], ["m", "dot"])
```
