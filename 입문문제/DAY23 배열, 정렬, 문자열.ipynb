## 특이한 정렬

```python
# 정수 n을 기준으로 n과 가까운 수부터 정렬하려고 합니다.
# 이때 n으로부터의 거리가 같다면 더 큰 수를 앞에 오도록 배치합니다.
# 정수가 담긴 배열 numlist와 정수 n이 주어질 때 numlist의 원소를 n으로부터 가까운 순서대로 정렬한 배열을
# return하도록 solution 함수를 완성해주세요.
def solution(numlist, n):
    return sorted(numlist, key=lambda x: (abs(x - n), -x))


# 설명
def solution(numlist, n):                 # numlist를 n 기준으로 정렬하는 함수 정의
    return sorted(                       # 정렬된 새로운 리스트를 반환
        numlist,                         # 정렬 대상 리스트
        key=lambda x: (                  # 각 원소 x에 대해 정렬 기준을 반환하는 함수
            abs(x - n),                  # x와 n의 거리 → 가까울수록 먼저 정렬
            -x                            # 거리가 같으면 x가 큰 값이 먼저 오도록 설정
        )                                # 튜플 기준: 앞 요소부터 순서대로 비교
    )                                    # sorted 함수 종료

# 다른 사람 풀이
def solution(numlist, n):
    answer = []
    if n in numlist:
        answer.append(n)
    for i in range(1,10000):
        if n + i in numlist:
            answer.append(n + i)

        if n - i in numlist:
            answer.append(n - i)

    return answer
```

```python
solution([1, 2, 3, 4, 5, 6], 4)
```

## 등수 매기기

```python
# 영어 점수와 수학 점수의 평균 점수를 기준으로 학생들의 등수를 매기려고 합니다. 
# 영어 점수와 수학 점수를 담은 2차원 정수 배열 score가 주어질 때,
# 영어 점수와 수학 점수의 평균을 기준으로 매긴 등수를 담은 배열을 return하도록 solution 함수를 완성해주세요.
def solution(score):                                   # 학생들의 점수 리스트를 받아 등수를 계산하는 함수
    avg = [(s[0] + s[1]) / 2 for s in score]            # 각 학생의 (영어+수학)/2 평균 점수를 리스트로 생성
    sorted_avg = sorted(avg, reverse=True)             # 평균 점수를 높은 점수부터 내림차순 정렬
    answer = [sorted_avg.index(a) + 1 for a in avg]    # 각 평균이 정렬된 리스트에서 몇 번째인지 찾아 등수 계산
                                                       # index는 0부터 시작하므로 +1
    return answer                                      # 학생 순서대로 등수 리스트 반환


# 다른 사람 풀이
def solution(score):
    a = sorted([sum(i) for i in score], reverse = True)
    return [a.index(sum(i))+1 for i in score]
```

```python
solution([[80, 70], [90, 50], [40, 70], [50, 80]])
```

## 옹알이 (1)

```python
# 머쓱이는 태어난 지 6개월 된 조카를 돌보고 있습니다.
# 조카는 아직 "aya", "ye", "woo", "ma" 네 가지 발음을 최대 한 번씩 사용해 조합한(이어 붙인) 발음밖에 하지 못합니다.
# 문자열 배열 babbling이 매개변수로 주어질 때, 머쓱이의 조카가 발음할 수 있는 단어의 개수를 return하도록
#  solution 함수를 완성해주세요.
def solution(babbling):                       # 발음 가능한 단어 개수를 세는 함수
    can = ["aya", "ye", "woo", "ma"]          # 허용된 발음 목록
    answer = 0                                # 정답(가능한 단어 수) 초기화
    
    for word in babbling:                     # babbling 리스트에서 단어 하나씩 순회
        temp = word                           # 원본 단어를 보존하기 위해 복사
        
        for c in can:                         # 허용된 발음을 하나씩 검사
            temp = temp.replace(c, " ", 1)    # 해당 발음이 있으면 공백으로 한 번만 치환

        if temp.strip() == "":                # 모든 발음이 제거되어 공백만 남았는지 확인
            answer += 1                       # 발음 가능하면 카운트 증가
    
    return answer                             # 최종 발음 가능한 단어 개수 반환

# 다른 사람 풀이
import re

def solution(babbling):
    regex = re.compile('^(aya|ye|woo|ma)+$')
    cnt=0
    for e in babbling:
        if regex.match(e):
            cnt+=1
    return cnt
```

```python
solution(["aya", "yee", "u", "maa", "wyeoo"])
```

## 로그인 성공?

```python
# 머쓱이는 프로그래머스에 로그인하려고 합니다. 
# 머쓱이가 입력한 아이디와 패스워드가 담긴 배열 id_pw와 회원들의 정보가 담긴 2차원 배열 db가 주어질 때, 
# 다음과 같이 로그인 성공, 실패에 따른 메시지를 return하도록 solution 함수를 완성해주세요.
# 아이디와 비밀번호가 모두 일치하는 회원정보가 있으면 "login"을 return합니다.
# 로그인이 실패했을 때 아이디가 일치하는 회원이 없다면 “fail”를, 
# 아이디는 일치하지만 비밀번호가 일치하는 회원이 없다면 “wrong pw”를 return 합니다.
def solution(id_pw, db):                     # 로그인 판별을 수행하는 함수 정의
    input_id, input_pw = id_pw               # 입력값 id_pw에서 아이디와 비밀번호를 분리
    
    for member_id, member_pw in db:          # DB에 저장된 (아이디, 비밀번호) 쌍을 하나씩 순회
        if input_id == member_id:            # 입력한 아이디가 DB의 아이디와 일치하는지 확인
            if input_pw == member_pw:        # 비밀번호까지 일치하는 경우
                return "login"               # 로그인 성공 반환
            else:                            # 아이디는 맞지만 비밀번호가 틀린 경우
                return "wrong pw"            # 비밀번호 오류 반환
    
    return "fail"                            # DB에 아이디 자체가 없는 경우 실패 반환

# 다른 사람 풀이
def solution(id_pw, db):                              # 로그인 판별 함수 정의
    if db_pw := dict(db).get(id_pw[0]):               # db를 dict로 변환 후 입력 아이디로 비밀번호 조회
                                                        # 동시에 조회 결과를 db_pw 변수에 저장
        return "login" if db_pw == id_pw[1] else "wrong pw"
                                                        # DB 비밀번호와 입력 비밀번호 비교
                                                        # 같으면 "login", 다르면 "wrong pw" 반환
    return "fail"                                     # 아이디가 DB에 없으면 실패 반환
```

```python
solution(["meosseugi", "1234"], [["rardss", "123"], ["yyoom", "1234"], ["meosseugi", "1234"]])
```
