## 문자열 밀기

```python
# 문자열 "hello"에서 각 문자를 오른쪽으로 한 칸씩 밀고
# 마지막 문자는 맨 앞으로 이동시키면 "ohell"이 됩니다.
# 이것을 문자열을 민다고 정의한다면
# 문자열 A와 B가 매개변수로 주어질 때,
# A를 밀어서 B가 될 수 있다면 밀어야 하는 최소 횟수를 return하고
# 밀어서 B가 될 수 없으면 -1을 return 하도록 solution 함수를 완성해보세요.
def solution(A, B):                         # 문자열 A를 오른쪽으로 회전해서 B가 되는 최소 횟수 반환
    if A == B:                              # A와 B가 처음부터 같은 경우
        return 0                            # 회전할 필요 없으므로 0 반환
    
    for i in range(1, len(A)+1):            # 1회전부터 문자열 길이만큼 반복
        A = A[-1] + A[:-1]                  # A의 마지막 문자를 맨 앞으로 이동 (오른쪽 회전 1회)
        if A == B:                          # 회전 결과가 B와 같은지 확인
            return i                        # 같다면 현재 회전 횟수 반환

    return -1                               # 끝까지 못 찾으면 -1 반환


# 다른 사람 풀이
from collections import deque  # 양쪽 끝에서 빠르게 회전/추가/삭제 가능한 자료구조

def solution(A, B):
    a, b = deque(A), deque(B)   # 문자열 A, B를 각각 deque로 변환
                               # (문자 하나하나를 원소로 갖는 큐)

    for cnt in range(0, len(A)):  # 0번부터 최대 len(A)-1번까지 밀어보기
        if a == b:                # 현재 A 상태가 B와 같다면
            return cnt            # 지금까지 민 횟수(cnt) 반환

        a.rotate(1)               # deque를 오른쪽으로 1칸 회전
                                  # 예: abc -> cab

    return -1                     # 끝까지 밀어도 B가 안 나오면 -1 반환

# 다른 사람 풀이
solution=lambda a,b:(b*2).find(a)
```

## 종이 자르기

```python
# 머쓱이는 큰 종이를 1 x 1 크기로 자르려고 합니다. 
# 예를 들어 2 x 2 크기의 종이를 1 x 1 크기로 자르려면 
# 최소 가위질 세 번이 필요합니다.
# 정수 M, N이 매개변수로 주어질 때, 
# M x N 크기의 종이를 최소로 가위질 해야하는 횟수를 
# return 하도록 solution 함수를 완성해보세요.
def solution(M, N):          # M x N 크기의 초콜릿을 1칸씩 나누기 위한 최소 쪼개기 횟수 계산
    return M * N - 1         # 전체 조각 수(M*N)에서 처음 한 덩어리를 제외한 나머지 개수만큼 쪼개기 필요

# 다른 사람 풀이
def get_cut_cnt_dfs(width, height):
    # 두 변 중 작은 값을 width, 큰 값을 height로 정렬
    # (항상 같은 기준으로 자르기 위해)
    width, height = min(width, height), max(width, height)

    # 종이가 1 x 1 크기라면
    # 이미 목표 크기이므로 더 이상 자를 필요가 없음
    if width == 1 and height == 1:
        return 0

    # 한 번 가위질을 한다 → +1
    # height를 반으로 잘라 두 개의 종이로 나눔
    # 1) width x (height//2)
    # 2) width x (height - height//2)
    # 두 조각에 대해 다시 같은 작업을 재귀적으로 수행
    return (
        1
        + get_cut_cnt_dfs(width, height // 2)
        + get_cut_cnt_dfs(width, height - height // 2)
    )


def solution(M, N):
    # M x N 크기의 종이를
    # 1 x 1 크기가 될 때까지 자르는데 필요한 최소 가위질 횟수 계산
    return get_cut_cnt_dfs(M, N)
```

```python
solution(2, 2)
```

## 연속된 수의 합

```python
# 연속된 세 개의 정수를 더해 12가 되는 경우는 3, 4, 5입니다.
# 두 정수 num과 total이 주어집니다.
# 연속된 수 num개를 더한 값이 total이 될 때,
# 정수 배열을 오름차순으로 담아 return하도록
# solution함수를 완성해보세요.

# 전체 합으로 중심 위치 계산
# 거기서 왼쪽 끝(start) 찾기
# start부터 num개 숫자 1씩 증가시키며 생성
def solution(num, total):                                  # 연속된 num개의 정수 합이 total이 되도록 하는 함수
    start = total / num - (num - 1) / 2                    # 시작값 = 평균 - (개수-1)/2 # 가운데 - 거리
    return [int(start + i) for i in range(num)]            # 시작값부터 num개를 1씩 증가시켜 리스트 생성


def solution(num, total):
    return [(total - (num * (num - 1) // 2)) // num + i for i in range(num)]

def solution(num, total):
    for start in range(-1000, 1001):
        s = 0
        arr = []
        
        for i in range(num):
            arr.append(start + i)
            s += start + i
        
        if s == total:
            return arr
```

## 다음에 올 숫자

```python
# 등차수열 혹은 등비수열 common이 매개변수로 주어질 때,
# 마지막 원소 다음으로 올 숫자를 return 하도록 solution 함수를 완성해보세요.
def solution(common):                                    # 등차수열 또는 등비수열 다음 항 계산 함수 정의
    if common[1] - common[0] == common[2] - common[1]:   # 앞 두 차로 등차인지 판단
        return common[-1] + (common[1] - common[0])     # 등차면 마지막 항에 공차를 더해서 다음 항 반환
    else:                                                # 등차가 아니면 등비라고 가정
        return common[-1] * (common[1] // common[0])     # 등비면 마지막 항에 공비 곱해서 다음 항 반환


# 다른 사람 풀이
def solution(common):              # common이라는 리스트(수열)를 입력으로 받는 함수 정의             
    a, b, c = common[:3]            # 수열의 처음 3개 값을 각각 a, b, c에 저장
    if (b - a) == (c - b):          # 앞의 두 차이와 뒤의 두 차이가 같다면 → 등차수열
        return common[-1] + (b - a) # 마지막 값에 공차를 더해 다음 항을 반환
    else:                           # 등차수열이 아니라면 → 등비수열
        return common[-1] * (b // a) # 마지막 값에 공비를 곱해 다음 항을 반환
```
