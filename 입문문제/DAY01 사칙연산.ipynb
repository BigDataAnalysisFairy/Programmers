## 두 수의 합 구하기

```python
# 정수 num1과 num2가 주어질 때, 
# num1과 num2의 합을 return하도록
# soltuion 함수를 완성해주세요.
def solution(num1, num2):
    return num1 + num2

# 다른 사람 풀이
solution=lambda * x :sum(x)

solution=lambda num1, num2 : num1 + num2
```

```python
solution(2, 3)
```

## 두 수의 차 구하기

```python
# 정수 num1과 num2가 주어질 때, 
# num1에서 num2를 뺀 값을 return하도록
# soltuion 함수를 완성해주세요.
def solution(num1, num2):
    return num1 - num2

# 다른 사람 풀이
solution = lambda num1, num2 : num1 - num2
```

```python
solution(2, 3)
```

## 두 수의 곱 구하기

```python
# 정수 num1, num2가 매개변수 주어집니다. 
# num1과 num2를 곱한 값을 return 하도록 
# solution 함수를 완성해주세요.
def solution(num1, num2):
    return num1 - num2

# 다른 사람 풀이
solution = lambda num1, num2 : num1 * num2
```

```python
solution(2, 3)
```

## 몫 구하기

```python
# 정수 num1, num2가 매개변수로 주어질 때, 
# num1을 num2로 나눈 몫을 return 하도록 
# solution 함수를 완성해주세요.
def solution(num1, num2):
    return num1 // num2

# 다른 사람 풀이
solution = lambda num1, num2 : num1 // num2
```

```python
solution(10, 5)
```
