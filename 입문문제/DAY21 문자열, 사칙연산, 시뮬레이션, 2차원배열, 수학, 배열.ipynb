## 숨어있는 숫자의 덧셈 (2)

```python
# 문자열 my_string이 매개변수로 주어집니다.
# my_string은 소문자, 대문자, 자연수로만 구성되어있습니다.
# my_string안의 자연수들의 합을 return하도록 solution 함수를 완성해주세요.
def solution(my_string):
    num = '' # True
    total = 0

    for i in my_string:
        if i.isdigit(): # i 가 숫자면
            num += i # num에 i 더해줌
        else: # i가 알파벳이고
            if num: # num이 채워져 있으면
                total += int(num) # total에 더하기
                num = ''

    if num: # 반복문이 끝나고 num에 남은 숫자가 있다면
        total += int(num) # total에 더하기
    return total

# 다른 사람 풀이
def solution(my_string):
    s = ''.join(i if i.isdigit() else ' ' for i in my_string)
    return sum(int(i) for i in s.split())
```

## 안전지대

```python
# 다음 그림과 같이 지뢰가 있는 지역과 지뢰에 인접한 
# 위, 아래, 좌, 우 대각선 칸을 모두 위험지역으로 분류합니다.
# 지뢰는 2차원 배열 board에 1로 표시되어 있고 board에는 
# 지뢰가 매설 된 지역 1과, 지뢰가 없는 지역 0만 존재합니다.
# 지뢰가 매설된 지역의 지도 board가 매개변수로 주어질 때, 
# 안전한 지역의 칸 수를 return하도록 solution 함수를 완성해주세요.
def solution(board):
    n = len(board)
    danger = [[0] * n for _ in range(n)]
    
    dx = [-1, -1, -1, 0, 0, 1, 1, 1]
    dy = [-1, 0, 1, -1, 1, -1, 0, 1]
    
    for i in range(n):
        for j in range(n):
            if board[i][j] == 1:
                danger[i][j] = 1
                for k in range(8):
                    ni = i + dx[k]
                    nj = j + dy[k]
                    if 0 <= ni < n and 0 <= nj < n:
                        danger[ni][nj] = 1
    
    answer = 0
    for i in range(n):
        for j in range(n):
            if danger[i][j] == 0:
                answer += 1
                
    return answer

# 다른 사람 풀이
def solution(board):                                # 안전한 지역 개수를 구하는 함수
    n = len(board)                                  # 보드의 크기 (n x n)

    danger = [[0] * n for _ in range(n)]             # 위험 지역 표시용 2차원 배열 (0 = 안전)

    directions = [                                   # 지뢰 기준 이동 방향 (8방향 + 자기 자신)
        (-1, -1), (-1, 0), (-1, 1),                  # 위쪽 3방향
        (0, -1),  (0, 0),  (0, 1),                   # 같은 줄 (왼쪽, 자기 자신, 오른쪽)
        (1, -1),  (1, 0),  (1, 1)                    # 아래쪽 3방향
    ]

    for i in range(n):                               # 행 순회
        for j in range(n):                           # 열 순회
            if board[i][j] == 1:                     # 현재 칸이 지뢰라면
                for dx, dy in directions:            # 주변 8방향 + 자기 자신 확인
                    nx = i + dx                      # 이동한 행 좌표
                    ny = j + dy                      # 이동한 열 좌표

                    if 0 <= nx < n and 0 <= ny < n:  # 보드 범위 안인지 확인
                        danger[nx][ny] = 1           # 해당 칸을 위험 지역으로 표시

    safe_count = 0                                   # 안전 지역 개수 저장 변수

    for i in range(n):                               # danger 배열 전체 순회
        for j in range(n):
            if danger[i][j] == 0:                    # 위험하지 않은 칸이면
                safe_count += 1                      # 안전 지역 개수 증가

    return safe_count                                # 최종 안전 지역 개수 반환

# 다른 사람 풀이
def solution(board):
    n = len(board)
    danger = set()
    for i, row in enumerate(board):
        for j, x in enumerate(row):
            if not x:
                continue
            danger.update((i+di, j+dj) for di in [-1,0,1] for dj in [-1, 0, 1])
    return n*n - sum(0 <= i < n and 0 <= j < n for i, j in danger)
```

## 삼각형의 완성조건 (2)

```python
# 선분 세 개로 삼각형을 만들기 위해서는 다음과 같은 조건을 만족해야 합니다.
# 가장 긴 변의 길이는 다른 두 변의 길이의 합보다 작아야 합니다.
# 삼각형의 두 변의 길이가 담긴 배열 sides이 매개변수로 주어집니다. 나머지 한 변이 될 수 있는 정수의 개수를
# return하도록 solution 함수를 완성해주세요.
def solution(sides):                     # 두 변이 주어졌을 때 가능한 나머지 한 변의 개수를 구하는 함수
    a, b = sorted(sides)                 # 두 변을 오름차순 정렬 (a ≤ b로 만들기)
    max_c = a + b - 1                    # 삼각형 조건상 가능한 가장 큰 c (c < a + b)
    min_c = b - a + 1                    # 삼각형 조건상 가능한 가장 작은 c (b < a + c → c > b - a)
    return max_c - min_c + 1             # 가능한 c의 개수 = 끝값 - 시작값 + 1


# 다른 사람 풀이
def solution(sides):
    return sum(sides) - max(sides) + min(sides) - 1
```

## 외계어 사전

```python
# PROGRAMMERS-962 행성에 불시착한 우주비행사 머쓱이는 
# 외계행성의 언어를 공부하려고 합니다.
# 알파벳이 담긴 배열 spell과 외계어 사전 dic이 매개변수로 주어집니다. 
# spell에 담긴 알파벳을 한번씩만 모두 사용한 단어가 dic에 존재한다면 1, 
# 존재하지 않는다면 2를 return하도록 solution 함수를 완성해주세요.
def solution(spell, dic):                      # spell: 사용할 문자 리스트, dic: 단어 리스트
    for i in dic:                              # dic에 있는 단어를 하나씩 꺼냄 (i는 문자열)
        lst = []                               # 현재 단어의 문자들을 담을 빈 리스트 생성
        for j in i:                            # 단어 i의 각 문자를 하나씩 순회
            lst.append(j)                      # 문자를 리스트 lst에 추가
        if sorted(spell) == sorted(lst):       # spell과 현재 단어의 문자 구성을 정렬해서 비교
            return 1                           # 완전히 같은 문자 구성이라면 1 반환
    return 2                                   # 어떤 단어도 조건을 만족하지 못하면 2 반환


# 다른 사람 풀이
def solution(spell, dic):                     # spell: 필요한 문자들, dic: 검사할 단어 목록
    spell = set(spell)                        # spell을 집합으로 변환 (중복 제거 + 빠른 비교)
    for s in dic:                             # dic에 있는 단어를 하나씩 확인
        if not spell - set(s):                # spell에 있는데 s에는 없는 문자가 하나도 없으면(공집합이면 : False) 
            return 1                          # 해당 단어로 spell을 모두 만들 수 있으므로 1 반환
    return 2                                  # 어떤 단어도 조건을 만족하지 못하면 2 반환
```
