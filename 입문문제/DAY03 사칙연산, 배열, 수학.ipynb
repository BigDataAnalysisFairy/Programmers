## 나머지 구하기

```python
# 정수 num1, num2가 매개변수로 주어질 때, 
# num1를 num2로 나눈 나머지를 return 하도록
#  solution 함수를 완성해주세요.
def solution(num1, num2):
    return num1 % num2

# 다른 사람 풀이
solution = lambda num1, num2 : num1 % num2
```

```python
solution(3, 2)
```

## 중앙값 구하기

```python
# 중앙값은 어떤 주어진 값들을 크기의 순서대로 정렬했을 때 
# 가장 중앙에 위치하는 값을 의미합니다.
# 예를 들어 1, 2, 7, 10, 11의 중앙값은 7입니다. 
# 정수 배열 array가 매개변수로 주어질 때, 중앙값을 return 하도록 
# solution 함수를 완성해보세요.
def solution(array):
    return sorted(array)[len(array) // 2]

# 다른 사람 풀이
def solution(array):
    array.sort()
    return array[len(array)//2]
```

```python
solution([1, 2, 7, 10, 11])
```

## 최빈값 구하기

```python
# 최빈값은 주어진 값 중에서 가장 자주 나오는 값을 의미합니다. 
# 배열 array가 매개변수로 주어질 때, 최빈값을 return 하도록 
# solution 함수를 완성해보세요. 최빈값이 여러 개면 -1을 return 합니다.
def solution(array):                          # 배열에서 최빈값(가장 많이 나온 값)을 구하는 함수
    counts = []                               # 각 값과 그 등장 횟수를 저장할 리스트

    for x in set(array):                      # 중복을 제거한 값들만 하나씩 순회
        counts.append([x, array.count(x)])   # [값, 해당 값의 등장 횟수] 형태로 저장

    counts.sort(key=lambda x: x[1], reverse=True)  # 등장 횟수(x[1]) 기준으로 내림차순 정렬

    if len(counts) > 1 and counts[0][1] == counts[1][1]:  
        return -1                             # 최빈값이 2개 이상이면 -1 반환

    return counts[0][0]                       # 최빈값(가장 많이 등장한 값) 반환


# 다른 사람 풀이
def solution(array):
    while len(array) != 0:
        for i, a in enumerate(set(array)):
            array.remove(a)
        if i == 0: return a
    return -1
```

```python
solution([1, 2, 3, 3, 3, 4])
```

## 짝수는 싫어요

```python
# 정수 n이 매개변수로 주어질 때,
# n 이하의 홀수가 오름차순으로 담긴 배열을 return하도록
# solution 함수를 완성해주세요.
def solution(n):
    lst = []
    for i in range(1, n+1):
        if i % 2 == 1:
            lst.append(i)
    return lst

# 다른 사람 풀이
def solution(n):
    return [i for i in range(1, n+1, 2)]
```

```python
solution(10)
```
