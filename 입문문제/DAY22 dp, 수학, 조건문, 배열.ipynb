## 저주의 숫자 3

```python
# 3x 마을 사람들은 3을 저주의 숫자라고 생각하기 때문에
# 3의 배수와 숫자 3을 사용하지 않습니다.
# 3x 마을 사람들의 숫자는 다음과 같습니다.
# 정수 n이 매개변수로 주어질 때, n을 3x 마을에서 사용하는 숫자로 바꿔
# return하도록 solution 함수를 완성해주세요.
def solution(n):                     # n번째로 3이 들어가지 않는 수를 구하는 함수 정의
    count = 0                         # 조건을 만족한 숫자의 개수를 세는 변수
    i = 1                             # 1부터 숫자를 하나씩 검사하기 위한 변수
    
    while True:                       # 원하는 n번째 숫자를 찾을 때까지 무한 반복
        if i % 3 != 0 and '3' not in str(i):  # i가 3의 배수가 아니고, 숫자에 '3'이 포함되지 않으면
            count += 1                # 조건을 만족했으므로 count를 1 증가
        
        if count == n:                # 조건을 만족한 숫자의 개수가 n개가 되면
            return i                  # 그때의 숫자 i를 반환하고 함수 종료
        
        i += 1                        # 다음 숫자를 확인하기 위해 i를 1 증가


# 다른 사람 풀이
# 1을 더 더하는 느낌으로
def solution(n):
    answer = 0
    for _ in range(n):
        answer += 1
        while answer % 3 == 0 or '3' in str(answer):
            answer += 1
    return answer
```

## 평행

```python
# 점 네 개의 좌표를 담은 이차원 배열  dots가 다음과 같이 매개변수로 주어집니다.
# [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]
# 주어진 네 개의 점을 두 개씩 이었을 때, 
# 두 직선이 평행이 되는 경우가 있으면 1을 없으면 0을 
# return 하도록 solution 함수를 완성해보세요.
def solution(dots):                         # 네 개의 점 좌표가 담긴 리스트 dots를 입력으로 받는 함수
    a, b, c, d = dots                       # dots에 들어있는 네 점을 각각 a, b, c, d로 분리

    def is_parallel(p1, p2, p3, p4):        # 두 선분 (p1-p2), (p3-p4)가 평행한지 판별하는 함수
        return (p2[1] - p1[1]) * (p4[0] - p3[0]) == (p4[1] - p3[1]) * (p2[0] - p1[0])
                                            # 첫 번째 선분의 y차이 × 두 번째 선분의 x차이
                                            # 두 번째 선분의 y차이 × 첫 번째 선분의 x차이
                                            # → 두 기울기가 같은지 교차곱으로 비교

    if is_parallel(a, b, c, d):              # 선분 a-b 와 c-d 가 평행한 경우
        return 1                             # 평행한 쌍이 있으므로 1 반환

    if is_parallel(a, c, b, d):              # 선분 a-c 와 b-d 가 평행한 경우
        return 1                             # 평행한 쌍이 있으므로 1 반환

    if is_parallel(a, d, b, c):              # 선분 a-d 와 b-c 가 평행한 경우
        return 1                             # 평행한 쌍이 있으므로 1 반환

    return 0                                 # 어떤 경우에도 평행하지 않으면 0 반환
```

```python
solution([[1, 4], [9, 2], [3, 8], [11, 6]])
```

## 겹치는 선분의 길이

```python
# # 선분 3개가 평행하게 놓여 있습니다.
# 세 선분의 시작과 끝 좌표가 [[start, end], [start, end], [start, end]] 형태로
# 들어있는 2차원 배열 lines가 매개변수로 주어질 때, 
# 두 개 이상의 선분이 겹치는 부분의 길이를 return 하도록 solution 함수를 완성해보세요.
# lines가 [[0, 2], [-3, -1], [-2, 1]]일 때 그림으로 나타내면 다음과 같습니다.
# 선분이 두 개 이상 겹친 곳은 [-2, -1], [0, 1]로 길이 2만큼 겹쳐있습니다.
def solution(lines):                          # 3개의 선분 정보가 들어있는 lines를 입력으로 받는 함수
    answer = 0                                # 두 개 이상 선분이 겹치는 구간의 전체 길이를 저장할 변수

    for x in range(-100, 100):                # 좌표 -100부터 99까지를 1단위씩 순회 (문제 좌표 범위)
        cnt = 0                               # 현재 위치 x를 포함하는 선분의 개수 카운트

        for start, end in lines:              # 각 선분의 시작점(start)과 끝점(end)을 하나씩 확인
            if start <= x < end:              # x가 선분 [start, end) 구간 안에 포함되면
                cnt += 1                      # 해당 선분이 x를 포함하므로 카운트 증가

        if cnt >= 2:                          # x 위치를 지나는 선분이 2개 이상이면
            answer += 1                       # 겹치는 길이 1만큼 추가

    return answer                             # 두 개 이상 선분이 겹친 전체 길이 반환

# 다른 사람 풀이
def solution(lines):
    answer = 0                                # 최종 겹치는 길이

    for i in range(3):                        # 첫 번째 선분 선택
        for j in range(i + 1, 3):             # 두 번째 선분 선택
            s1, e1 = lines[i]                 # 첫 번째 선분의 시작, 끝
            s2, e2 = lines[j]                 # 두 번째 선분의 시작, 끝

            overlap = min(e1, e2) - max(s1, s2)  # 겹치는 구간 길이 계산
            if overlap > 0:                   # 실제로 겹칠 때만
                answer += overlap             # 겹친 길이 누적

    return answer

# 다른 사람 풀이
def solution(lines):
    sets = [set(range(min(l), max(l))) for l in lines]
    return len(sets[0] & sets[1] | sets[0] & sets[2] | sets[1] & sets[2])
```

```python
solution([[0, 1], [2, 5], [3, 9]])
```

## 유한소수 판별하기

```python
# 소수점 아래 숫자가 계속되지 않고 유한개인 소수를 유한소수라고 합니다. 
# 분수를 소수로 고칠 때 유한소수로 나타낼 수 있는 분수인지 판별하려고 합니다. 
# 유한소수가 되기 위한 분수의 조건은 다음과 같습니다.
# 기약분수로 나타내었을 때, 분모의 소인수가 2와 5만 존재해야 합니다.
# 두 정수 a와 b가 매개변수로 주어질 때, a/b가 유한소수이면 1을, 
# 무한소수라면 2를 return하도록 solution 함수를 완성해주세요.
import math                      # 최대공약수(gcd)를 쓰기 위한 math 모듈 import

def solution(a, b):
    g = math.gcd(a, b)           # a와 b의 최대공약수 계산
    b //= g                      # 분모 b를 최대공약수로 나눠 기약분수로 만듦

    while b % 2 == 0:            # 분모에 2가 더 이상 안 남을 때까지
        b //= 2                  # 2로 계속 나눔

    while b % 5 == 0:            # 분모에 5가 더 이상 안 남을 때까지
        b //= 5                  # 5로 계속 나눔

    return 1 if b == 1 else 2    # 분모가 1이면 유한소수(1), 아니면 무한소수(2)
```

```python
solution(7, 20)
```
