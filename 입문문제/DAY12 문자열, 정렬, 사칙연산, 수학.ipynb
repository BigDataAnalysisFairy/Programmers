## 모음 제거

```python
# 영어에선 a, e, i, o, u 다섯 가지 알파벳을 모음으로 분류합니다.
# 문자열 my_string이 매개변수로 주어질 때 모음을 제거한 문자열을 return하도록
# solution 함수를 완성해주세요.
def solution(my_string):
    vowel = ['a', 'e', 'i', 'o', 'u']
    ans = ''
    for i in my_string:
        if i in vowel:
            pass
        else:
            ans += i
    return ans

# 다른 사람 풀이
def solution(my_string):
    return "".join([i for i in my_string if not(i in "aeiou")])
```

```python
solution('bus')
```

## 문자열 정렬하기(1)

```python
# 문자열 my_string이 매개변수로 주어질 때, 
# my_string 안에 있는 숫자만 골라 오름차순 정렬한 리스트를
# return 하도록 solution 함수를 작성해보세요.
def solution(my_string):
    lst = []
    for i in my_string:
        if i.isnumeric():
            lst.append(int(i))
    lst.sort()
    return lst

# 다른 사람 풀이
def solution(my_string):
    return sorted([int(c) for c in my_string if c.isdigit()])
```

```python
solution("hi12392")
```

## 숨어있는 숫자의 덧셈 (1)

```python
# 문자열 my_string이 매개변수로 주어집니다. 
# my_string안의 모든 자연수들의 합을
# return하도록 solution 함수를 완성해주세요.
def solution(my_string):
    ans = 0
    for i in my_string:
        if i.isdigit():
            ans += int(i)
    return ans

# 다른 사람 풀이
def solution(my_string):
    return sum(int(i) for i in my_string if i.isdigit())
```

```python
solution("aAb1B2cC34oOp")
```

## 소인수분해

```python
# 소인수분해란 어떤 수를 소수들의 곱으로 표현하는 것입니다.
# 예를 들어 12를 소인수 분해하면 2 * 2 * 3 으로 나타낼 수 있습니다.
# 따라서 12의 소인수는 2와 3입니다.
# 자연수 n이 매개변수로 주어질 때 n의 소인수를 오름차순으로 담은 배열을 return하도록
# solution 함수를 완성해주세요.

def solution(n):              # 자연수 n을 입력으로 받는 함수
    answer = []               # 중복 없는 소인수들을 저장할 리스트
    i = 2                     # 가장 작은 소수 2부터 나누기 시작
    
    while i * i <= n:         # i가 n의 제곱근 이하일 동안만 검사 (시간 최적화)
        if n % i == 0:        # i로 나누어 떨어지면 (i는 소인수)
            answer.append(i)  # 소인수 i를 결과 리스트에 추가
            
            while n % i == 0: # i가 더 이상 나누어지지 않을 때까지
                n //= i       # n을 i로 계속 나눠서 중복 제거 (중복된 소인수 제거 목적)
        
        i += 1                # 다음 숫자를 소인수 후보로 검사
    
    if n > 1:                 # √n까지 나눈 후에도 n이 1보다 크면
        answer.append(n)      # 남은 n은 소수이므로 소인수로 추가
    
    return answer             # 소인수 리스트 반환


# 다른 사람 풀이
def solution(n):
    answer = []
    d = 2
    while d <= n:
        if n % d == 0:
            n /= d
            if d not in answer:
                answer.append(d)
        else:
            d += 1
    return answer
```
