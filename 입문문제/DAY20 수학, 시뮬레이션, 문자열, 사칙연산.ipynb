## 직사각형 넓이 구하기

```python
# 2차원 좌표 평면에 변이 축과 평행한 직사각형이 있습니다.
# 직사각형 네 꼭짓점의 좌표
# [[x1, y1], [x2, y2], [x3, y3], [x4, y4]]가
# 담겨있는 배열 dots가 매개변수로 주어질 때, 
# 직사각형의 넓이를 return 하도록 solution 함수를 완성해보세요.
def solution(dots):
    i = []
    j = []

    for x, y in dots:
        i.append(x)
        j.append(y)
        
    return (max(i) - min(i)) * (max(j) - min(j))

# 다른 사람 풀이
def solution(dots):
    return (max(dots)[0] - min(dots)[0]) * (max(dots)[1] - min(dots)[1])
```

```python
solution([[1, 1], [2, 1], [2, 2], [1, 2]])
```

## 캐릭터의 좌표

```python
# 머쓱이는 RPG게임을 하고 있습니다. 
# 게임에는 up, down, left, right 방향키가 있으며
# 각 키를 누르면 위, 아래, 왼쪽, 오른쪽으로 한 칸씩 이동합니다.
# 예를 들어 [0,0]에서 up을 누른다면 캐릭터의 좌표는 [0, 1], down을 누른다면 [0, -1],
# left를 누른다면 [-1, 0], right를 누른다면 [1, 0]입니다. 
# 머쓱이가 입력한 방향키의 배열 keyinput와 맵의 크기 board이 매개변수로 주어집니다.
# 캐릭터는 항상 [0,0]에서 시작할 때 키 입력이 모두 끝난 뒤에 캐릭터의 좌표 [x, y]를
# return하도록 solution 함수를 완성해주세요.
# [0, 0]은 board의 정 중앙에 위치합니다.
# 예를 들어 board의 가로 크기가 9라면 캐릭터는 왼쪽으로 최대 [-4, 0]까지
# 오른쪽으로 최대 [4, 0]까지 이동할 수 있습니다.
def solution(keyinput, board):
    ans = [0, 0]

    for i in keyinput:
        if i == 'left':
            ans[0] -= 1
        elif i == 'right':
            ans[0] += 1
        elif i == 'up':
            ans[1] += 1
        elif i == 'down':
            ans[1] -= 1

        if ans[0] > ((board[0]-1) // 2):
            ans[0] = ((board[0]-1) // 2)
        elif ans[0] < -((board[0]-1) // 2):
            ans[0] = -((board[0]-1) // 2)
        elif ans[1] > ((board[1]-1) // 2):
            ans[1] = ((board[1]-1) // 2)
        elif ans[1] < -((board[1]-1) // 2):
            ans[1] = -((board[1]-1) // 2)

    return ans

# 다른 사람 풀이
def solution(keyinput, board):
    x_lim, y_lim = board[0]//2, board[1]//2
    move = {'left':(-1,0),'right':(1,0),'up':(0,1),'down':(0,-1)}
    x,y = 0,0
    for k in keyinput:
        dx,dy = move[k]
        if abs(x+dx)>x_lim or abs(y+dy)>y_lim:
            continue
        else:
            x,y = x+dx,y+dy

    return [x,y]

# GPT
def solution(keyinput, board):
    x, y = 0, 0                             # 현재 캐릭터 좌표 초기화 [0,0]
    
    max_x = board[0] // 2                    # 오른쪽/왼쪽 이동 최대 범위 계산
    max_y = board[1] // 2                    # 위/아래 이동 최대 범위 계산
    
    for key in keyinput:                     # 입력된 방향키 하나씩 처리
        if key == "up" and y < max_y:       # "up" 키이고 위쪽 범위를 벗어나지 않으면
            y += 1                           # y 좌표 1 증가
        elif key == "down" and y > -max_y:  # "down" 키이고 아래쪽 범위를 벗어나지 않으면
            y -= 1                           # y 좌표 1 감소
        elif key == "left" and x > -max_x:  # "left" 키이고 왼쪽 범위를 벗어나지 않으면
            x -= 1                           # x 좌표 1 감소
        elif key == "right" and x < max_x:  # "right" 키이고 오른쪽 범위를 벗어나지 않으면
            x += 1                           # x 좌표 1 증가
    
    return [x, y]                            # 최종 좌표 반환
```

## 최댓값 만들기 (2)

```python
# 정수 배열 numbers가 매개변수로 주어집니다. 
# numbers의 원소 중 두 개를 곱해 만들 수 있는 최댓값을
# return하도록 solution 함수를 완성해주세요.
def solution(numbers):
    numbers = sorted(numbers)
    return max(numbers[0] * numbers[1], numbers[-1]*numbers[-2])

# 다른 사람 풀이
def solution(numbers):
    numbers.sort()
    return max(numbers[0] * numbers[1], numbers[-1] * numbers[-2])
```

```python
solution([1, 2, -3, 4, -5])
```

## 다항식 더하기

```python
# 한 개 이상의 항의 합으로 이루어진 식을 다항식이라고 합니다. 
# 다항식을 계산할 때는 동류항끼리 계산해 정리합니다. 
# 덧셈으로 이루어진 다항식 polynomial이 매개변수로 주어질 때, 
# 동류항끼리 더한 결괏값을 문자열로 return 하도록 solution 함수를 완성해보세요. 
# 같은 식이라면 가장 짧은 수식을 return 합니다.
def solution(polynomial):
    total_x = []
    total_num = []
    
    for i in polynomial.split():
        if i == '+':
            continue
        elif 'x' in i:
            coef = i.replace('x', '')
            total_x.append(1 if coef == '' else int(coef))
        else:
            total_num.append(int(i))
    
    x_sum = sum(total_x)
    num_sum = sum(total_num)

    if x_sum == 0:
        return str(num_sum)
    
    if num_sum == 0:
        return 'x' if x_sum == 1 else f'{x_sum}x'
    
    x_part = 'x' if x_sum == 1 else f'{x_sum}x'
    return f'{x_part} + {num_sum}'

# 다른 사람 풀이
def solution(polynomial):
    # polynomial 문자열을 " + " 기준으로 나누어 항들의 리스트로 만듦
    terms = polynomial.split(" + ")  # 예: "3x + 7 + x" -> ['3x', '7', 'x']
    
    x_total = 0  # x항들의 계수를 합산할 변수
    num_total = 0  # 상수항들의 합을 저장할 변수

    # 각 항을 순회하며 x항과 상수항 구분
    for term in terms:
        if 'x' in term:  # 항에 'x'가 있으면 x항
            if term == 'x':  # 계수가 생략된 경우, 즉 'x'만 있는 경우
                x_total += 1  # 계수 1을 더함
            else:
                x_total += int(term[:-1])  # '3x' 같은 경우 마지막 문자 'x' 제거 후 int로 변환하여 합산
        else:  # 'x'가 없으면 상수항
            num_total += int(term)  # 상수값을 정수로 변환하여 합산

    # 최종 결과 문자열 초기화
    result = ''

    # x항이 존재하면 먼저 처리
    if x_total > 0:
        if x_total == 1:  # 계수가 1이면 생략하고 'x'만 사용
            result += 'x'
        else:  # 계수가 1보다 크면 '계수x' 형태로 추가
            result += str(x_total) + 'x'

    # 상수항이 존재하면 처리
    if num_total > 0:
        if result != '':  # 이미 x항이 존재하면 " + " 추가
            result += ' + '
        result += str(num_total)  # 상수항 값을 문자열로 추가

    return result  # 최종 결과 문자열 반환
```

```python
solution("3x + 7 + x")
```
