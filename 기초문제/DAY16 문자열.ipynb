## 배열에서 문자열 대소문자 변환하기

```python
# 문자열 배열 strArr가 주어집니다.
# 모든 원소가 알파벳으로만 이루어져 있을 때,
# 배열에서 홀수번째 인덱스의 문자열은 모든 문자를 대문자로,
# 짝수번째 인덱스의 문자열은 모든 문자를 소문자로 바꿔서 반환하는 solution 함수를 완성해 주세요.
def solution(strArr):
    s = []
    for i in range(len(strArr)):
        if i % 2 == 1:
            s.append(strArr[i].upper())
        else:
            s.append(strArr[i].lower())
    return s

# 다른 사람 풀이
def solution(strArr):
    answer = []
    for i in range(len(strArr)):
        if i % 2 == 0:
            answer.append(strArr[i].lower())
        else:
            answer.append(strArr[i].upper())
    return answer
```

```python
solution(["AAA","BBB","CCC","DDD"])
```

## A 강조하기

```python
# 문자열 myString이 주어집니다.
# myString에서 알파벳 "a"가 등장하면 전부 "A"로 변환하고, 
# "A"가 아닌 모든 대문자 알파벳은 소문자 알파벳으로 변환하여 
# return 하는 solution 함수를 완성하세요.
def solution(myString):
    s = ''
    for i in myString:
        if i == 'a':
            s += 'A'
        elif i.isupper() and i != 'a' and i != 'A':
            s += i.lower()
        else:
            s += i
    return s


# 다른 사람 풀이
def solution(myString):
    return myString.lower().replace('a', 'A')

def solution(myString):
    answer = ''
    answer = myString.lower()
    answer = answer.replace('a', 'A')
    return answer
```

```python
solution("PrOgRaMmErS")
```

## 특정한 문자를 대문자로 바꾸기

```python
# 영소문자로 이루어진 문자열 my_string과 영소문자 1글자로 이루어진 문자열 alp가 
# 매개변수로 주어질 때, my_string에서 alp에 해당하는
# 모든 글자를 대문자로 바꾼 문자열을 return 하는 solution 함수를 작성해 주세요.
def solution(my_string, alp):
    if alp in my_string:
        return my_string.replace(my_string[my_string.find(alp)], alp.upper())
    else:
        return my_string
    
# 다른 사람 풀이
def solution(my_string, alp):
    return my_string.replace(alp, alp.upper())
```

```python
solution("programmers", 'p')
```
