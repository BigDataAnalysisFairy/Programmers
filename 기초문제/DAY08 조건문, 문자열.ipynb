## 간단한 논리 연산

```python
# boolean 변수 x1, x2, x3, x4가 매개변수로 주어질 때, 
# 다음의 식의 true/false를 return 하는 solution 함수를 작성해 주세요.
# v = 둘 중 하나가 t면 t임
# ^ = 둘 중 하나가 F면 f임
def solution(x1, x2, x3, x4):
    return (x1 or x2) and (x3 or x4)

# 다른 사람 풀이
def solution(x1, x2, x3, x4):
    return (x1 | x2) & (x3 | x4)
```

```python
# 1부터 6까지 숫자가 적힌 주사위가 네 개 있습니다.
# 네 주사위를 굴렸을 때 나온 숫자에 따라 다음과 같은 점수를 얻습니다.
# 네 주사위에서 나온 숫자가 모두 p로 같다면 1111 × p점을 얻습니다.
# 세 주사위에서 나온 숫자가 p로 같고
# 나머지 다른 주사위에서 나온 숫자가 q(p ≠ q)라면 (10 × p + q)2 점을 얻습니다.
# 주사위가 두 개씩 같은 값이 나오고,
# 나온 숫자를 각각 p, q(p ≠ q)라고 한다면 (p + q) × |p - q|점을 얻습니다.
# 어느 두 주사위에서 나온 숫자가 p로 같고
# 나머지 두 주사위에서 나온 숫자가 각각 p와 다른 q, r(q ≠ r)이라면 q × r점을 얻습니다.
# 네 주사위에 적힌 숫자가 모두 다르다면 나온 숫자 중 가장 작은 숫자 만큼의 점수를 얻습니다.
# 네 주사위를 굴렸을 때 나온 숫자가 정수 매개변수 a, b, c, d로 주어질 때,
# 얻는 점수를 return 하는 solution 함수를 작성해 주세요.
from collections import Counter

def solution(a, b, c, d):
    nums = [a, b, c, d]
    cnt = Counter(nums)  # 각 숫자가 몇 번 나왔는지 세기
    items = list(cnt.items())  # [(숫자, 개수), (숫자, 개수), ...]

    # 1️⃣ 네 개 다 같은 경우
    if len(cnt) == 1:
        p = nums[0]
        return 1111 * p

    # 2️⃣ 세 개 같은 경우
    elif len(cnt) == 2:
        (p, pc), (q, qc) = items
        # p가 3번, q가 1번 나왔을 때
        if pc == 3:
            return (10 * p + q) ** 2
        # q가 3번, p가 1번 나왔을 때
        elif qc == 3:
            return (10 * q + p) ** 2
        # 둘 다 2번씩 나왔을 때
        else:
            return (p + q) * abs(p - q)

    # 3️⃣ 하나만 2번, 나머지 2개 다 다름
    elif len(cnt) == 3:
        # 2번 나온 숫자 p 찾기
        for num, c in cnt.items():
            if c == 2:
                p = num
            else:
                pass
        # 1번씩 나온 숫자 q, r
        q, r = [num for num, c in cnt.items() if c == 1]
        return q * r

    # 4️⃣ 모두 다름
    else:
        return min(nums)
        
# 다른 사람 풀이
def solution(a, b, c, d):
    nums = [a, b, c, d]
    counts = [nums.count(i) for i in nums]
    if max(counts) == 4:
        return a * 1111
    elif max(counts) == 3:
        p = nums[counts.index(3)]
        q = nums[counts.index(1)]
        return (10 * p + q) ** 2
    elif max(counts) == 2:
        if min(counts) == 2:
            return (a + c) * abs(a - c) if a == b else (a + b) * abs(a - b)
        else:
            p = nums[counts.index(2)]
            return (a * b * c * d) / p**2
    else:
        return min(nums)
    
# GPT
from collections import Counter  # Counter: 리스트에서 각 숫자의 등장 횟수를 세어주는 자료구조

def solution(a, b, c, d):
    dice = [a, b, c, d]            # 주사위 숫자들을 리스트로 모음, 반복문과 최소값 계산에 편리
    counts = Counter(dice)         # 각 숫자가 몇 번 나왔는지 세어서 딕셔너리 형태로 저장

    if len(counts) == 1:           # 서로 다른 숫자가 1개 → 네 개 모두 같은 경우
        p = dice[0]                # 모든 숫자가 같으므로 첫 번째 숫자를 p로 사용
        return 1111 * p            # 점수 공식: 1111 * p

    elif len(counts) == 2:         # 서로 다른 숫자가 2개 → 패턴 3+1 또는 2+2
        for num, cnt in counts.items():  # 각 숫자와 그 등장 횟수를 확인
            if cnt == 3:           # 숫자가 3번 등장하면 → 세 개 같은 숫자 p
                p = num
            else:                  # 나머지 하나는 q
                q = num
        if max(counts.values()) == 3:    # 최대 등장 횟수가 3 → 3+1 패턴
            return (10 * p + q) ** 2    # 점수 공식: (10 * p + q) ^ 2
        else:                            # 두 쌍이 각각 2개씩 → 2+2 패턴
            p, q = counts.keys()         # 서로 다른 두 숫자를 p, q로 저장
            return (p + q) * abs(p - q)  # 점수 공식: (p + q) * |p - q|

    elif len(counts) == 3:         # 서로 다른 숫자가 3개 → 한 쌍만 같고 나머지 두 개는 다름 (2+1+1 패턴)
        for num, cnt in counts.items():  # 각 숫자와 그 등장 횟수를 확인
            if cnt == 2:           # 두 번 나온 숫자 → 한 쌍 p
                p = num
            else:                  # 한 번 나온 숫자 두 개 → q와 r
                if 'q' not in locals():  # q가 아직 정의되지 않았다면
                    q = num             # 첫 번째 1개짜리 숫자를 q로 저장
                else:                     # q가 이미 정의되어 있다면
                    r = num             # 두 번째 1개짜리 숫자를 r로 저장
        return q * r                # 점수 공식: 한 쌍 제외한 나머지 두 숫자 곱

    else:                           # 서로 다른 숫자가 4개 → 모두 다름
        return min(dice)            # 점수 공식: 가장 작은 숫자가 점수
```

## 글자 이어 붙여 문자열 만들기

```python
# 문자열 my_string과 정수 배열 index_list가 매개변수로 주어집니다. 
# my_string의 index_list의 원소들에 해당하는 인덱스의 글자들을 순서대로 이어 붙인 문자열을 
# return 하는 solution 함수를 작성해 주세요.
def solution(my_string, index_list):
    a = ''
    for i in index_list:
        a += my_string[i]
    return a

# 다른 사람 풀이
def solution(my_string, index_list):
    return ''.join([my_string[idx] for idx in index_list])
```

```python
solution("cvsgiorszzzmrpaqpe", [16, 6, 5, 3, 12, 14, 11, 11, 17, 12, 7])
```

## 9로 나눈 나머지

```python
# 음이 아닌 정수를 9로 나눈 나머지는 
# 그 정수의 각 자리 숫자의 합을 9로 나눈 나머지와 같은 것이 알려져 있습니다.
# 이 사실을 이용하여 음이 아닌 정수가 문자열 number로 주어질 때, 
# 이 정수를 9로 나눈 나머지를 return 하는 
# solution 함수를 작성해주세요.
def solution(number):
    a = 0
    for i in number:
        a += int(i)
    return a % 9

# 다른 사람 풀이
def solution(number):
    return int(number) % 9
```

```python
solution("123")
```

# 문자열 여러 번 뒤집기

```python
# 문자열 my_string과 이차원 정수 배열 queries가 매개변수로 주어집니다.
# queries의 원소는 [s, e] 형태로, my_string의 인덱스 s부터 인덱스 e까지를 뒤집으라는 의미입니다.
# my_string에 queries의 명령을 순서대로 처리한 후의 문자열을 return 하는 solution 함수를 작성해 주세요.
def solution(my_string, queries):
    s = list(my_string)  # 문자열을 리스트로 변환 (문자 하나하나 접근 가능)

    for start, end in queries: # 슬라이싱으로 해당 구간을 뒤집어서 다시 대입
        s[start:end+1] = reversed(s[start:end+1])

    return ''.join(s)  # 리스트를 다시 문자열로 합쳐서 반환

# 다른 사람 풀이
def solution(my_string, queries):
    answer=list(my_string)
    for s,e in queries:
        answer[s:e+1]=answer[s:e+1][::-1]
    return ''.join(answer)
```

```python
solution("rermgorpsam", [[2, 3], [0, 7], [5, 9], [6, 10]])
```
