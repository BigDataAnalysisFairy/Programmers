## 정수를 나선형으로 배치하기

```python
# 양의 정수 n이 매개변수로 주어집니다. 
# n × n 배열에 1부터 n2 까지 정수를 인덱스 [0][0]부터 시계방향 나선형으로 배치한
# 이차원 배열을 return 하는 solution 함수를 작성해 주세요.
def solution(n):                              # n x n 달팽이 배열을 만드는 함수
    arr = [[0] * n for _ in range(n)]         # 모든 값이 0인 2차원 배열 생성
    
    dx = [0, 1, 0, -1]                        # 행 이동: 오른쪽, 아래, 왼쪽, 위
    dy = [1, 0, -1, 0]                        # 열 이동: 오른쪽, 아래, 왼쪽, 위
    
    x, y = 0, 0                               # 시작 위치 (0행 0열)
    d = 0                                    # 현재 방향 인덱스 (0 = 오른쪽)
    
    for num in range(1, n * n + 1):           # 1부터 n*n까지 숫자 채우기
        arr[x][y] = num                       # 현재 위치에 숫자 저장
        
        nx = x + dx[d]                        # 다음 행 위치 계산
        ny = y + dy[d]                        # 다음 열 위치 계산
        
        # 다음 위치가 범위를 벗어나거나 이미 값이 있으면
        if nx < 0 or nx >= n or ny < 0 or ny >= n or arr[nx][ny] != 0: # 위쪽 벽, 아래쪽 벽, 왼쪽 벽, 오른쪽 벽, 이미 채워진 칸 (막힌 곳)
            d = (d + 1) % 4                   # 방향을 시계방향으로 변경
            nx = x + dx[d]                    # 변경된 방향으로 다음 행 계산
            ny = y + dy[d]                    # 변경된 방향으로 다음 열 계산
        
        x, y = nx, ny                         # 실제 이동
    
    return arr                                # 완성된 달팽이 배열 반환

# 다른 사람 풀이
def solution(n):
    answer = [[None for j in range(n)] for i in range(n)]
    move = [[0, 1], [1, 0], [0, -1], [-1, 0]]
    x, y, m = 0, 0, 0
    for i in range(1, n**2 + 1):
        answer[y][x] = i
        if y + move[m][0] >= n or x + move[m][1] >= n or answer[y + move[m][0]][x + move[m][1]]:
            m = (m + 1) % len(move)
        y, x = y + move[m][0], x + move[m][1]
    return answer
```

## 특별한 이차원 배열2

```python
# n × n 크기의 이차원 배열 arr이 매개변수로 주어질 때,
# arr이 다음을 만족하면 1을 아니라면 0을 return 하는
# solution 함수를 작성해 주세요.
# 0 ≤ i, j < n인 정수 i, j에 대하여 arr[i][j] = arr[j][i]
def solution(arr):                                 # 함수 solution 정의, arr는 2차원 배열
    a = ''                                         # 비교 결과를 저장할 빈 문자열 a 초기화
    for i in range(len(arr)):                      # 행 인덱스 i를 0부터 arr 길이-1까지 반복
        for j in range(len(arr)):                  # 열 인덱스 j를 0부터 arr 길이-1까지 반복
            if arr[i][j] == arr[j][i]:            # arr[i][j]와 arr[j][i] 값이 같으면
                a += '1'                           # 문자열 a에 '1' 추가
            else:                                  # 그렇지 않으면
                a += '0'                           # 문자열 a에 '0' 추가
    if '0' in a:                                   # a 안에 '0'이 하나라도 있으면 (대칭이 아닌 값이 있으면)
        return 0                                   # 0 반환 (대칭 아님)                                          # 그렇지 않으면 (모든 값이 대칭)
    return 1                                       # 그렇지 않으면 1 반환 (대칭)

    
# 다른 사람 풀이
def solution(arr):
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i][j] != arr[j][i]:
                return 0
    return 1
```

## 정사각형으로 만들기

```python
# 이차원 정수 배열 arr이 매개변수로 주어집니다.
# arr의 행의 수가 더 많다면 열의 수가 행의 수와 같아지도록 각 행의 끝에 0을 추가하고,
# 열의 수가 더 많다면 행의 수가 열의 수와 같아지도록
# 각 열의 끝에 0을 추가한 이차원 배열을 return 하는 solution 함수를 작성해 주세요.
def solution(arr):                           # 함수 solution 정의, arr은 2차원 리스트
    rows = len(arr)                          # arr의 행(row) 개수 계산
    cols = len(arr[0])                       # 첫 번째 행의 길이 = 열(column) 개수 계산
    
    if rows > cols:                          # 행이 열보다 많으면
        for i in range(rows):                # 각 행에 대해 반복
            arr[i].extend([0] * (rows - cols))  # 행 뒤에 0을 추가하여 열 개수를 맞춤
    
    elif rows < cols:                        # 행이 열보다 적으면
        for _ in range(cols - rows):         # 부족한 행 개수만큼 반복
            arr.append([0] * cols)           # 길이가 cols인 새 행을 추가
    
    return arr                                # 정사각형 배열로 바뀐 arr 반환

# 다른 사람 풀이
def solution(arr):
    n=len(arr)
    m=len(arr[0])
    if n>m:
        for i in range(n):
            for j in range(n-m):
                arr[i].append(0)
    else:
        for i in range(m-n):
            arr.append([0]*m)

    return arr
```

## 이차원 배열 대각선 순회하기

```python
# 2차원 정수 배열 board와 정수 k가 주어집니다.
# i + j <= k를 만족하는 모든 (i, j)에 대한 board[i][j]의 합을 return 하는
# solution 함수를 완성해 주세요.
def solution(board, k):
    c = 0
    for i in range(len(board)): # 행
        for j in range(len(board[i])): # 열
            if i + j <= k:
                c+= board[i][j]      
    return c

# 다른 사람 풀이
def solution(board, k):
    answer = 0
    n, m = len(board), len(board[0])

    for i in range(n):
        for j in range(m):
            if i + j <= k:
                answer += board[i][j]
            else:
                break
    return answer
```

```python
solution([[0, 1, 2],[1, 2, 3],[2, 3, 4],[3, 4, 5]], 2)
```
