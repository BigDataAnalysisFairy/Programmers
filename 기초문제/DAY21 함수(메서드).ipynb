## 뒤에서 5등 위로

```python
# 정수로 이루어진 리스트 num_list가 주어집니다.
# num_list에서 가장 작은 5개의 수를 제외한 수들을 오름차순으로 담은 리스트를
# return하도록 solution 함수를 완성해주세요.
def solution(num_list):
    num_list.sort()
    return num_list[5:]

# 다른 사람 풀이
def solution(num_list):
    return sorted(num_list)[5:]
```

```python
solution([12, 4, 15, 46, 38, 1, 14, 56, 32, 10])
```

## 전국 대회 선발 고사

```python
# 0번부터 n - 1번까지 n명의 학생 중 3명을 선발하는 전국 대회 선발 고사를 보았습니다. 
# 등수가 높은 3명을 선발해야 하지만, 
# 개인 사정으로 전국 대회에 참여하지 못하는 학생들이 있어
# 참여가 가능한 학생 중 등수가 높은 3명을 선발하기로 했습니다.
# 각 학생들의 선발 고사 등수를 담은 정수 배열 rank와 
# 전국 대회 참여 가능 여부가 담긴 boolean 배열 attendance가 매개변수로 주어집니다.
# 전국 대회에 선발된 학생 번호들을 등수가 높은 순서대로 각각 a, b, c번이라고 할 때
# 10000 × a + 100 × b + c를 return 하는 solution 함수를 작성해 주세요.
def solution(rank, attendance):            # rank와 attendance 리스트를 입력으로 받는 함수
    candidates = []                        # (등수, 인덱스)를 저장할 빈 리스트 생성

    for i in range(len(rank)):             # rank의 인덱스를 0부터 끝까지 순회
        if attendance[i]:                  # i번째 학생이 참석(attendance[i] == True)이라면
            candidates.append((rank[i], i))# (해당 학생의 등수, 인덱스 i)를 튜플로 추가

    candidates.sort()                      # 등수(rank)를 기준으로 오름차순 정렬

    a = candidates[0][1]                   # 가장 등수가 높은 학생의 인덱스
    b = candidates[1][1]                   # 두 번째로 높은 학생의 인덱스
    c = candidates[2][1]                   # 세 번째로 높은 학생의 인덱스

    return 10000 * a + 100 * b + c          # 문제 조건에 맞게 계산한 값 반환

# 다른 사람 풀이
def solution(rank, attendance):
    arr = sorted([(x, i) for i, x in enumerate(rank) if attendance[i]])
    return arr[0][1] * 10000 + arr[1][1] * 100 + arr[2][1]
```

```python
solution([3, 7, 2, 5, 4, 6, 1], [0, 1, 0, 1, 1, 0, 0])
```

## 정수 부분

```python
# 실수 flo가 매개 변수로 주어질 때, flo의 정수 부분을 return하도록
# solution 함수를 완성해주세요.
def solution(flo):
    return int(flo)

# 다른 사람 풀이
def solution(flo):
    return flo // 1
```

```python
solution(1.42)
```

## 문자열 정수의 합

```python
# 한 자리 정수로 이루어진 문자열 num_str이 주어질 때, 
# 각 자리수의 합을 return하도록 
# solution 함수를 완성해주세요.
def solution(num_str):
    a = 0
    for i in num_str:
        a += int(i)
    return a

# 다른 사람 풀이
def solution(num_str):
    return sum(map(int, list(num_str)))
```

```python
solution('123456789')
```

## 문자열을 정수로 변환하기

```python
# 숫자로만 이루어진 문자열 n_str이 주어질 때, 
# n_str을 정수로 변환하여 return하도록 solution 함수를 완성해주세요.
def solution(n_str):
    return int(n_str)
```

```python
solution("10")
```
