## 커피 심부름

```python
# 팀의 막내인 철수는 아메리카노와 카페 라테만 판매하는 카페에서 
# 팀원들의 커피를 사려고 합니다. 
# 아메리카노와 카페 라테의 가격은 
# 차가운 것과 뜨거운 것 상관없이 각각 4500, 5000원입니다. 
# 각 팀원에게 마실 메뉴를 적어달라고 하였고, 
# 그 중에서 메뉴만 적은 팀원의 것은 차가운 것으로 통일하고
# "아무거나"를 적은 팀원의 것은 차가운 아메리카노로 통일하기로 하였습니다.
# 각 직원이 적은 메뉴가 문자열 배열 order로 주어질 때,
# 카페에서 결제하게 될 금액을 return 하는 solution 함수를 작성해주세요. 
# order의 원소는 아래의 것들만 들어오고, 각각의 의미는 다음과 같습니다.
# order의 원소 의미
# "iceamericano", "americanoice" : 차가운 아메리카노
# "hotamericano", "americanohot" : 따뜻한 아메리카노
# "icecafelatte", "cafelatteice" : 차가운 카페 라테
# "hotcafelatte", "cafelattehot" :따뜻한 카페 라테
# "americano" : 아메리카노
# "cafelatte" : 카페 라테
# "anything" : 아무거나

def solution(order):
    a = 0
    for i in order:
        if i == "iceamericano" or i == "americanoice":
            a += 4500
        elif i == "hotamericano" or i == "americanohot":
            a += 4500
        elif i == "icecafelatte" or i == "cafelatteice":
            a += 5000
        elif i == "hotcafelatte" or i == "cafelattehot":
            a += 5000
        elif i == "americano" or i == "anything":
            a += 4500
        elif i == 'cafelatte':
            a += 5000
    return a

# 다른 사람 풀이
def solution(order):
    answer = 0
    for want in order:
        if 'latte' in want:
            answer += 500
        answer += 4500

    return answer
```

```python
solution(["cafelatte", "americanoice", "hotcafelatte", "anything"])
```

## 그림 확대

```python
# 직사각형 형태의 그림 파일이 있고, 
# 이 그림 파일은 1 × 1 크기의 정사각형 크기의 픽셀로 이루어져 있습니다.
# 이 그림 파일을 나타낸 문자열 배열 picture과 정수 k가 매개변수로 주어질 때,
# 이 그림 파일을 가로 세로로 k배 늘린 그림 파일을 나타내도록
# 문자열 배열을 return 하는 solution 함수를 작성해 주세요.
def solution(picture, k):                 # picture(문자열 리스트)와 확대 비율 k를 받는 함수
    result = []                           # 최종 확대된 그림을 저장할 빈 리스트 생성
    
    for row in picture:                  # picture의 각 행(row)을 하나씩 순회
        expanded_row = ''.join(           # 가로로 k배 확대한 문자열 생성
            char * k for char in row      # row의 각 문자(char)를 k번 반복
        )
        
        for _ in range(k):                # 세로로 k배 늘리기 위해 k번 반복
            result.append(expanded_row)   # 가로 확대한 줄을 result에 추가
    
    return result                         # 최종 결과 반환

# 다른 사람 풀이
def solution(picture, k):
    answer = []
    for i in range(len(picture)):
        for _ in range(k):
            answer.append(picture[i].replace('.', '.' * k).replace('x', 'x' * k))
    return answer
```

```python
solution([".xx...xx.", "x..x.x..x", "x...x...x", ".x.....x.", "..x...x..", "...x.x...", "....x...."], 2)
```

## 조건에 맞게 수열 변환하기 3

```python
# 정수 배열 arr와 자연수 k가 주어집니다.
# 만약 k가 홀수라면 arr의 모든 원소에 k를 곱하고,
# k가 짝수라면 arr의 모든 원소에 k를 더합니다.
# 이러한 변환을 마친 후의 arr를 return 하는 solution 함수를 완성해 주세요.
def solution(arr, k):
    a = []
    for i in arr:
        if k % 2 == 1:
            a.append(i * k)
        else:
            a.append(i + k)
    return a

# 다른 사람 풀이
def solution(arr, k):
    return [i*k if k%2!=0 else i+k for i in arr]
```

```python
solution([1, 2, 3, 100, 99, 98], 3)
```

## I로 만들기

```python
# 알파벳 소문자로 이루어진 문자열 myString이 주어집니다.
# 알파벳 순서에서 "l"보다 앞서는 모든 문자를 "l"로 바꾼 문자열을 return하는 
# solution 함수를 완성해 주세요.
def solution(myString):
    alp = 'abcdefghijk'
    for i in myString:
        if i in alp:
            myString = myString.replace(i, 'l')
    return myString

# 다른 사람 풀이
def solution(myString):
    return myString.translate(str.maketrans('abcdefghijk', 'lllllllllll'))
```

```python
solution("abcdevwxyz")
```

## 특별한 이차원 배열 1

```python
# 정수 n이 매개변수로 주어질 때, 
# 다음과 같은 n × n 크기의 이차원 배열 arr를 return 하는
# solution 함수를 작성해 주세요.
# arr[i][j] (0 ≤ i, j < n)의 값은 i = j라면 1, 아니라면 0입니다.
def solution(n):
    list = []
    for _ in range(n):
        list.append([0] * n)
    for _ in range(n):
        list[_][_] = 1
    return list

# 다른 사람 풀이
def solution(n):
    result = []
    for i in range(n):
        arr = [0] * n
        arr[i] = 1
        result.append(arr)
    return result
# 참고 
# 이렇게 하면 내부 리스트들이 동일한 객체를 참조하게 됨
# 즉, 하나의 리스트를 수정하면 모든 리스트가 함께 변경됨
# a = [ [0] * n ] * n
```

```python
solution(3)
```
