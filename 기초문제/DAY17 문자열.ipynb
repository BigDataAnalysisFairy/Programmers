## 특정 문자열로 끝나는 가장 긴 부분 문자열 찾기

```python
# 문자열 myString과 pat가 주어집니다. 
# myString의 부분 문자열중 pat로 끝나는 가장 긴 부분 문자열을 찾아서 return 하는
# solution 함수를 완성해 주세요.
def solution(myString, pat):                 # 문자열 myString, pat을 입력으로 받음
    idx = myString.rfind(pat)                # pat이 마지막으로 시작되는 인덱스를 찾음
    return myString[:idx + len(pat)]         # pat이 끝나는 지점까지 문자열을 잘라 반환

def solution(myString, pat):
    return myString[:len(myString) - myString[::-1].index(pat[::-1])]
```

```python
solution("AbCdEFG", "dE")
```

## 문자열이 몇 번 등장하는지 세기

```python
# 문자열 myString과 pat이 주어집니다.
# myString에서 pat이 등장하는 횟수를
# return 하는 solution 함수를 완성해 주세요.
# 슬라이딩 윈도우 기술
def solution(myString, pat):                         # 문자열 myString에서 pat이 몇 번 등장하는지 세는 함수
    answer = 0                                       # pat이 등장한 횟수를 저장할 변수
    for i in range(len(myString) - len(pat) + 1):   # pat이 시작할 수 있는 모든 인덱스 범위 순회
        if myString[i:i+len(pat)] == pat:            # i부터 pat 길이만큼 잘라서 pat과 비교
            answer += 1                              # 같다면 등장 횟수 1 증가
    return answer                                    # 총 등장 횟수 반환

def solution(myString, pat):
    answer = 0
    for i, x in enumerate(myString):
        if myString[i:].startswith(pat):
            answer += 1
    return answer
```

```python
solution('banana', 'ana')
```

## ad 제거하기

```python
# 문자열 배열 strArr가 주어집니다.
# 배열 내의 문자열 중 "ad"라는 부분 문자열을 포함하고 있는 모든 문자열을
# 제거하고 남은 문자열을 순서를 유지하여 배열로 return 하는
# solution 함수를 완성해 주세요.
def solution(strArr):
    a = []
    for i in strArr:
        if 'ad' in i:
            pass
        else:
            a.append(i)
    return a

# 다른 사람 풀이
def solution(strArr):
    answer = []
    for x in strArr:
        if 'ad' in x: continue
        answer.append(x)
    return answer
```

## 공백으로 구분하기1

```python
# 단어가 공백 한 개로 구분되어 있는 문자열 my_string이 매개변수로 주어질 때, 
# my_string에 나온 단어를 앞에서부터 순서대로 담은 문자열 배열을 return 하는
# solution 함수를 작성해 주세요.
def solution(my_string):
    return my_string.split(' ')
```

```python
solution('i love you')
```
