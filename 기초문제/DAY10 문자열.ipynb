## 문자열의 앞의 n글자

```python
# 문자열 my_string과 정수 n이 매개변수로 주어질 때, 
# my_string의 앞의 n글자로 이루어진 문자열을 return 하는
# solution 함수를 작성해 주세요.
def solution(my_string, n):
    return my_string[:n]
```

```python
solution("ProgrammerS123", 11)
```

## 접두사인지 확인하기

```python
# 어떤 문자열에 대해서 접두사는 특정 인덱스까지의 문자열을 의미합니다. 
# 예를 들어, "banana"의 모든 접두사는 "b", "ba", "ban", "bana", "banan", "banana"입니다.
# 문자열 my_string과 is_prefix가 주어질 때, 
# is_prefix가 my_string의 접두사라면 1을, 
# 아니면 0을 return 하는 solution 함수를 작성해 주세요.
def solution(my_string, is_prefix):
    if my_string[:len(is_prefix)] == is_prefix:
        return 1
    else:
        return 0

# 다른 사람 풀이
def solution(my_string, is_prefix):
    return int(my_string.startswith(is_prefix))
```

```python
solution('banana', 'ban')
```

## 문자열 뒤집기

```python
# 문자열 my_string과 정수 s, e가 매개변수로 주어질 때,
#  my_string에서 인덱스 s부터 인덱스 e까지를 뒤집은 문자열을 return 하는 solution 함수를 작성해 주세요.
def solution(my_string, s, e):
    return my_string[:s] + my_string[s:e+1][::-1] + my_string[e+1:]
```

```python
solution("Progra21Sremm3", 6, 12)
```

## 세로 읽기

```python
# 문자열 my_string과 두 정수 m, c가 주어집니다. 
# my_string을 한 줄에 m 글자씩 가로로 적었을 때 
# 왼쪽부터 세로로 c번째 열에 적힌 글자들을 문자열로 return 하는
# solution 함수를 작성해 주세요.
def solution(my_string, m, c):          # 문자열 my_string을 m개씩 잘라서 각 묶음의 c번째 문자를 모아 반환
    a = ''                              # m개씩 임시로 문자를 쌓아둘 문자열
    b = ''                              # 최종 결과를 저장할 문자열
    for i in range(0, len(my_string)):  # 문자열의 모든 인덱스를 0부터 끝까지 순회
        a += my_string[i]               # 현재 문자를 a에 하나씩 추가
        if len(a) == m:                 # a에 문자가 m개 모였다면
            b += a[c-1]                 # a의 c번째 문자(0-index라서 c-1)를 결과 b에 추가
            a = ''                      # 다음 묶음을 위해 a를 초기화
    return b                            # 완성된 결과 문자열 반환

# 다른 사람 풀이
def solution(s, m, c):
    return s[c-1::m]

# 참고
"ihrhbakrfpndopljhygc"[1::4]
```

```python
solution("ihrhbakrfpndopljhygc", 4, 2)
```

## qr code

```python
# 두 정수 q, r과 문자열 code가 주어질 때, 
# code의 각 인덱스를 q로 나누었을 때 
# 나머지가 r인 위치의 문자를 앞에서부터 순서대로 이어 붙인 문자열을 return 하는
# solution 함수를 작성해 주세요.

# 초반 풀이
def solution(q, r, code):
    a = ''
    for i in range(len(code)):
        if (code.index(code[i]) % q) == r:
            a += code[i]
    return a

# 나의 실수
# code.index(code[i])는 code[i]가 처음 등장하는 위치를 반환하므로, 
# 중복된 문자가 있을 경우 원하는 인덱스를 얻지 못할 수 있음.

# 해결
def solution(q, r, code):
    a = ''
    for i in range(len(code)):
        if i % q == r:
            a += code[i]
    return a

# 다른 사람 풀이
def solution(q, r, code):
    return code[r::q]

# 참고
# index = x * q + r 임을 알 수 있고, 
# 우리가 필요한 index는 [0q+r, 1q+r, 2q+r... xq+r]입니다. 
# 즉 [r, q+r, 2q+r... xq+r]임을 알 수있고, 
# r로 시작해서 q씩 커지는 index만 가져오면 됩니다 :)
```

```python
solution(3, 1, "qjnwezgrpirldywt")
```
