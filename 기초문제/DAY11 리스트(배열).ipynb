## 문자 개수 세기

```python
# 알파벳 대소문자로만 이루어진 문자열 my_string이 주어질 때, 
# my_string에서 'A'의 개수, my_string에서 'B'의 개수,..., my_string에서 'Z'의 개수, 
# my_string에서 'a'의 개수, my_string에서 'b'의 개수,..., my_string에서 'z'의 개수를 
# 순서대로 담은 길이 52의 정수 배열을 return 하는 solution 함수를 작성해 주세요.
def solution(my_string):                     # 문자열 my_string에서 알파벳 대소문자 개수를 세는 함수
    import string                            # 알파벳 상수를 쓰기 위한 모듈 import

    upper = string.ascii_uppercase           # 'A'부터 'Z'까지 문자열
    lower = string.ascii_lowercase           # 'a'부터 'z'까지 문자열

    counts = [0] * 52                        # 대문자 26 + 소문자 26 = 총 52칸 리스트 생성

    for i in my_string:                      # 문자열의 각 문자를 하나씩 순회
        if i in upper:                       # 현재 문자가 대문자라면
            idx = upper.index(i)             # 대문자에서의 위치(0~25)를 구함
            counts[idx] += 1                 # 해당 위치의 개수 증가
        elif i in lower:                     # 현재 문자가 소문자라면
            idx = lower.index(i)             # 소문자에서의 위치(0~25)를 구함
            counts[idx + 26] += 1             # 소문자는 뒤쪽(26~51)에 저장
    return counts                            # 알파벳 개수 리스트 반환


# 다른 사람 풀이
def solution(my_string):
    answer=[0]*52
    for x in my_string:
        if x.isupper():
            answer[ord(x)-65]+=1
        else:
            answer[ord(x)-71]+=1
    return answer

# 다른 사람 풀이
def solution(my_string):
    answer = []
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
    for i in alphabet:
        answer.append(len(my_string.split(i))-1)
    return answer
```

## 글자 지우기

```python
# 문자열 my_string과 정수 배열 indices가 주어질 때, 
# my_string에서 indices의 원소에 해당하는 인덱스의 글자를 지우고 이어 붙인 문자열을
# return 하는 solution 함수를 작성해 주세요.
def solution(my_string, indices):
    a = ''
    for i in range(len(my_string)):
        if i in indices:
            pass
        else:
            a += my_string[i]
    return a

# 다른 사람 풀이
def solution(my_string, indices):
    answer = ''
    for i in range(len(my_string)):
        if i not in indices:
            answer+=my_string[i]
    return answer
```

## 카운트 다운

```python
# 정수 start_num와 end_num가 주어질 때, 
# start_num에서 end_num까지 1씩 감소하는 수들을 
# 차례로 담은 리스트를 return하도록 solution 함수를 완성해주세요.
def solution(start_num, end_num):
    a = []
    for i in range(start_num,end_num-1,-1):
        a.append(i)
    return a

# 다른 사람 풀이
def solution(start, end):
    return list(range(start,end-1,-1))
```

```python
solution(10, 3)
```

## 가까운 1 찾기

```python
# 정수 배열 arr가 주어집니다. 이때 arr의 원소는 1 또는 0입니다. 
# 정수 idx가 주어졌을 때, idx보다 크면서 배열의 값이 1인 가장 작은 인덱스를 찾아서 반환하는 
# solution 함수를 완성해 주세요.
# 단, 만약 그러한 인덱스가 없다면 -1을 반환합니다.
def solution(arr, idx):
    for i in range(idx, len(arr)):
        if arr[i] == 1:
            return i
    return -1

# 다른 사람 풀이
def solution(arr, idx):
    answer = -1
    for i in range(idx,len(arr)):
        if arr[i]==1:
            answer = i
            break
    return answer
```

```python
solution([0, 0, 0, 1], 1)
```
