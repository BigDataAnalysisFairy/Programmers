## 배열의 길이를 2의 거듭제곱으로 만들기

```python
# 정수 배열 arr이 매개변수로 주어집니다.
# arr의 길이가 2의 정수 거듭제곱이 되도록 arr 뒤에 정수 0을 추가하려고 합니다.
# arr에 최소한의 개수로 0을 추가한 배열을 return 하는
# solution 함수를 작성해 주세요.
def solution(arr):                  # arr이라는 리스트를 입력으로 받는 함수 정의
    a = 1                            # a를 1로 초기화 (2의 거듭제곱을 만들기 위한 시작값)
    while a < len(arr):              # a가 arr 길이보다 작은 동안 반복
        a *= 2                       # a를 2배로 증가
        if a > len(arr):             # a가 arr 길이를 초과하면
            break                    # 반복문 탈출
    while True:                      # 무한 반복 시작
        if len(arr) < a:             # arr 길이가 a보다 작으면
            arr.append(0)            # arr 뒤에 0을 하나 추가
        if len(arr) == a:            # arr 길이가 a와 같아지면
            break                    # 무한 반복 종료
    return arr                       # 길이가 a가 된 arr 반환


# 다른 사람 풀이
def solution(arr):
    a = 1
    b = len(arr)
    while a < b :
        a *= 2
    return arr + [0] * (a-b)
```

```python
solution([1, 2, 3, 4, 5, 6])
```

## 배열 비교하기

```python
# 이 문제에서 두 정수 배열의 대소관계를 다음과 같이 정의
# 두 배열의 길이가 다르다면, 배열의 길이가 긴 쪽이 더 큽니다.
# 배열의 길이가 같다면 각 배열에 있는 모든 원소의 합을 비교하여 다르다면
# 더 큰 쪽이 크고, 같다면 같습니다.
# 두 정수 배열 arr1과 arr2가 주어질 때,
# 위에서 정의한 배열의 대소관계에 대하여 arr2가 크다면 -1,
# arr1이 크다면 1, 두 배열이 같다면 0을 return 하는
# solution 함수를 작성해 주세요.
def solution(arr1, arr2):
    if len(arr1) > len(arr2):
        return 1
    elif len(arr1) < len(arr2):
        return -1
    else:
        a = 0
        b = 0
        for i in arr1:
            a += i
        for j in arr2:
            b += j

        if a > b:
            return 1
        elif a == b:
            return 0
        else:
            return -1     

# 다른 사람 풀이
def solution(arr1, arr2):
    if len(arr1) < len(arr2) :
        return -1
    elif len (arr1) > len(arr2) :
        return 1
    elif sum(arr1) > sum(arr2):
        return 1
    elif sum(arr1) < sum(arr2):
        return -1
    else :
        return 0
```

```python
solution([49, 13], [70, 11, 2])
```

## 배열의 길이에 따라 다른 연산하기

```python
# 정수 배열 arr과 정수 n이 매개변수로 주어집니다.
# arr의 길이가 홀수라면 arr의 모든 짝수 인덱스 위치에 n을 더한 배열을
# arr의 길이가 짝수라면 arr의 모든 홀수 인덱스 위치에 n을 더한 배열을
# return 하는 solution 함수를 작성해 주세요.
def solution(arr, n):
    temp = []
    if len(arr) % 2 == 1:
        for i in range(len(arr)):
            if i % 2 == 0:
                arr[i] += n
                temp.append(arr[i])
            else:
                temp.append(arr[i])
        return temp
    else:
        for i in range(len(arr)):
            if i % 2 == 1:
                arr[i] += n
                temp.append(arr[i])
            else:
                temp.append((arr[i]))
        return temp

# 다른 사람 풀이
def solution(arr, n):
    N=len(arr) # 5
    if N%2: # 짝수
        for i in range(0,N,2): 
            arr[i]+=n
    else: # # 홀수
        for i in range(1,N,2): 
            arr[i]+=n
    return arr

# 재풀이
def solution(arr, n):
    if len(arr) % 2 == 1:
        for i in range(0, len(arr), 2):
            arr[i] += n
    else:
        for i in range(1, len(arr), 2):
            arr[i] += n
    return arr
```

## 문자열 묶기

```python
# 문자열 배열 strArr이 주어집니다.
# strArr의 원소들을 길이가 같은 문자열들끼리 그룹으로 묶었을 때
# 가장 개수가 많은 그룹의 크기를 return 하는 solution 함수를 완성해 주세요.
from collections import defaultdict          # 기본값을 자동으로 만들어주는 defaultdict 사용

def solution(strArr):                         # 문자열 배열 strArr를 입력으로 받는 함수
    length_count = defaultdict(int)           # key가 없으면 기본값 0을 가지는 딕셔너리 생성
    for s in strArr:                          # strArr 안의 각 문자열을 하나씩 순회
        length_count[len(s)] += 1             # 문자열 길이를 key로 해서 개수 1 증가
    return max(length_count.values())         # 가장 많이 등장한 길이의 개수 반환

# 다른 사람 풀이
def solution(strArr):
    a=[0]*31
    for x in strArr: 
        a[len(x)]+=1
    return max(a)
```

```python
solution(["a","bc","d","efg","hi"])
```

## 뒤에서 5등까지

```python
# 정수로 이루어진 리스트 num_list가 주어집니다. 
# num_list에서 가장 작은 5개의 수를 오름차순으로 담은 리스트를 return하도록 
# solution 함수를 완성해주세요.
def solution(num_list):
    num_list.sort()
    return num_list[:5]

# 다른 사람 풀이
def solution(num_list):
    return sorted(num_list)[:5]
```

```python
solution([12, 4, 15, 46, 38, 1, 14])
```
