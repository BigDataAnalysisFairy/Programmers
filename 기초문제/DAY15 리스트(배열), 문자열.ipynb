## 조건에 맞게 수열 변환하기 1

```python
# 정수 배열 arr가 주어집니다.
# arr의 각 원소에 대해 값이 50보다 크거나 같은 짝수라면 2로 나누고,
# 50보다 작은 홀수라면 2를 곱합니다.
# 그 결과인 정수 배열을 return 하는 solution 함수를 완성해 주세요.
def solution(arr):
    lst = []
    for i in arr:
        if i >= 50 and i % 2 == 0:
            lst.append(i // 2)
        elif i < 50 and i % 2 == 1:
            lst.append(i * 2)
        else:
            lst.append(i)
    return lst

# 다른 사람 풀이
def solution(arr):
    answer = []
    for i in arr:
        if i>=50 and i % 2 == 0:
            answer.append(i / 2)
        elif i<50 and i % 2 != 0:
            answer.append(i * 2)
        else:
            answer.append(i)
    return answer
```

```python
solution([1, 2, 3, 100, 99, 98])
```

## 조건에 맞게 수열 변환하기 2

```python
# 정수 배열 arr가 주어집니다. 
# arr의 각 원소에 대해 값이 50보다 크거나 같은 짝수라면 2로 나누고,
# 50보다 작은 홀수라면 2를 곱하고 다시 1을 더합니다.
# 이러한 작업을 x번 반복한 결과인 배열을 arr(x)라고 표현했을 때,
# arr(x) = arr(x + 1)인 x가 항상 존재합니다.
# 이러한 x 중 가장 작은 값을 return 하는 solution 함수를 완성해 주세요.
# 단, 두 배열에 대한 "="는 두 배열의 크기가 서로 같으며, 
# 같은 인덱스의 원소가 각각 서로 같음을 의미합니다.
# 배열이 변하지 않게 되는 최초의 반복 횟수를 구하는 문제
def solution(arr):                     # 함수 정의: 정수 배열 arr를 입력으로 받음
    count = 0                          # 배열이 변환된 횟수를 세는 변수, 처음엔 0
    while True:                        # 배열이 더 이상 변하지 않을 때까지 무한 반복
        next_arr = []                  # 이번 반복에서 새로 만들어질 배열
        for i in range(len(arr)):      # arr의 인덱스 0부터 끝까지 하나씩 순회
            if arr[i] >= 50 and arr[i] % 2 == 0:   # 값이 50 이상이면서 짝수라면
                next_arr.append(arr[i] // 2)      # 해당 값을 2로 나눈 값을 추가
            elif arr[i] < 50 and arr[i] % 2 == 1: # 값이 50 미만이면서 홀수라면
                next_arr.append(arr[i] * 2 + 1)  # 2를 곱하고 1을 더한 값 추가
            else:                                 # 위 두 조건에 모두 해당하지 않으면
                next_arr.append(arr[i])          # 값이 변하지 않으므로 그대로 추가
        if next_arr == arr:            # 변환 전 배열과 변환 후 배열이 완전히 같다면
            break                      # 더 이상 변하지 않으므로 반복 종료
        arr = next_arr                 # arr를 새 배열로 갱신
        count += 1                     # 변환이 한 번 일어났으므로 count 1 증가
    return count                       # 배열이 안정화되기까지 걸린 횟수 반환

# 다른 사람 풀이
def solution(arr):
    answer = 0
    old = arr
    while(True):
        new = []
        for i in old:
            if i>=50 and i%2 == 0:
                i = i/2
            elif i<50 and i%2 == 1:
                i = i*2 + 1
            new.append(int(i))
        if old == new:
            break
        else:
            old = new
            answer += 1
    return answer
```

```python
solution([1, 2, 3, 100, 99, 98])
```

## 1로 만들기

```python
# 정수가 있을 때, 짝수라면 반으로 나누고, 홀수라면 1을 뺀 뒤 반으로 나누면,
# 마지막엔 1이 됩니다. 예를 들어 10이 있다면 다음과 같은 과정으로 1이 됩니다.
# 10 / 2 = 5
# (5 - 1) / 2 = 2
# 2 / 2 = 1
# 위와 같이 3번의 나누기 연산으로 1이 되었습니다.
# 정수들이 담긴 리스트 num_list가 주어질 때,
# num_list의 모든 원소를 1로 만들기 위해서 필요한 나누기 연산의 횟수를
# return하도록 solution 함수를 완성해주세요.
def solution(num_list):                 # 정수 배열 num_list를 입력으로 받는 함수
    a = 0                               # 연산 횟수를 누적해서 저장할 변수
    for i in range(len(num_list)):      # num_list의 인덱스 0부터 끝까지 순회
        while num_list[i] != 1:         # 현재 원소가 1이 될 때까지 반복 # 기술!
            if num_list[i] % 2 == 0:    # 현재 원소가 짝수라면
                num_list[i] = num_list[i] // 2  # 2로 나눈 몫으로 값 변경
                a += 1                  # 연산을 한 번 했으므로 횟수 1 증가
            else:                       # 현재 원소가 홀수라면
                num_list[i] = (num_list[i] - 1) // 2  # 1을 빼서 짝수 만든 뒤 2로 나눔
                a += 1                  # 연산을 한 번 했으므로 횟수 1 증가
    return a                            # 모든 원소를 1로 만들기까지의 총 연산 횟수 반환


# 2번째 풀이
def solution(num_list):
    a = 0
    for i in num_list:
        while i != 1:
            if i % 2 == 0:
                i = i // 2
                a +=1
            else:
                i = (i - 1) //2
                a +=1
    return a
    
# 다른 사람 풀이
def solution(num_list):
    answer = 0

    for n in num_list:
        while n != 1:
            n //= 2
            answer += 1

    return answer
```

```python
solution([12, 4, 15, 1, 14])
```

## 길이에 따른 연산

```python
# 정수가 담긴 리스트 num_list가 주어질 때,
# 리스트의 길이가 11 이상이면 리스트에 있는 모든 원소의 합을
# 10 이하이면 모든 원소의 곱을 return하도록 solution 함수를 완성해주세요.
def solution(num_list):
    sum1 = 0
    sum2 = 1
    if len(num_list) >= 11:
        for i in num_list:
            sum1 += i
        return sum1
    else:
        for i in num_list:
            sum2 *= i
        return sum2

# 다른 사람 풀이
# prod : 리스트 안에 있는 모든 걸 곱하는 함수
from math import prod
def solution(num_list):
    return sum(num_list) if len(num_list)>=11 else prod(num_list)
```

## 원하는 문자열 찾기

```python
# 알파벳으로 이루어진 문자열 myString과 pat이 주어집니다. 
# myString의 연속된 부분 문자열 중 pat이 존재하면 1을 그렇지 않으면 0을 
# return 하는 solution 함수를 완성해 주세요.
# 단, 알파벳 대문자와 소문자는 구분하지 않습니다.
def solution(myString, pat):
    myString = myString.lower()
    pat = pat.lower()
    if pat in myString:
        return 1
    else:
        return 0

# 다른 사람 풀이
def solution(myString, pat):
    return int(pat.lower() in myString.lower())
```

```python
solution("AbCdEfG", "aBc")
```
